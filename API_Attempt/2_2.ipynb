```python
import pandas as pd
import requests
import time
from urllib.parse import quote
# Query Google Maps
from dotenv import load_dotenv
import os

# Load environment variables from .env file
load_dotenv()
PLACES_API_KEY = os.getenv("PLACES_API_KEY")  # This will get your API key from the environment
API_KEY=PLACES_API_KEY
if not API_KEY:
    raise ValueError("API_KEY not found. Please set it in your .env file.")
```

```python
# Google Places API Functions
def build_place_search_query(row):
    """Build a search query string optimized for Places API text search."""
    components = []
    
    # Business name components (chain and label)
    if pd.notna(row.get('chain')): 
        components.append(str(row['chain']))
    if pd.notna(row.get('label')): 
        components.append(str(row['label']))
    
    # Address components
    if pd.notna(row.get('address')): 
        components.append(str(row['address']))
    if pd.notna(row.get('city')): 
        components.append(str(row['city']))
    if pd.notna(row.get('state')): 
        components.append(str(row['state']))
    if pd.notna(row.get('zip_code')): 
        components.append(str(row['zip_code']))
    
    return ', '.join(components)

def get_place_details(place_id):
    """Get detailed information about a place using Place Details API.
    Returns phone numbers, address components, and other detailed info."""
    
    if not place_id:
        return {}
    
    # Fields we want to retrieve from Place Details API
    fields = [
        'name', 'formatted_address', 'address_components', 'geometry/location',
        'formatted_phone_number', 'international_phone_number', 'vicinity',
        'rating', 'types', 'business_status', 'website', 'opening_hours'
    ]
    
    url = f'https://maps.googleapis.com/maps/api/place/details/json?place_id={place_id}&fields={",".join(fields)}&key={API_KEY}'
    
    try:
        response = requests.get(url)
        data = response.json()
        
        if data['status'] == 'OK':
            result = data['result']
            
            # Extract address components into a structured format
            address_components = {}
            for component in result.get('address_components', []):
                types = component.get('types', [])
                long_name = component.get('long_name', '')
                short_name = component.get('short_name', '')
                
                # Map common address component types
                if 'street_number' in types:
                    address_components['street_number'] = long_name
                elif 'route' in types:
                    address_components['street_name'] = long_name
                elif 'locality' in types:
                    address_components['city'] = long_name
                elif 'administrative_area_level_1' in types:
                    address_components['state'] = short_name
                    address_components['state_long'] = long_name
                elif 'postal_code' in types:
                    address_components['postal_code'] = long_name
                elif 'country' in types:
                    address_components['country'] = short_name
                    address_components['country_long'] = long_name
                elif 'administrative_area_level_2' in types:
                    address_components['county'] = long_name
                elif 'sublocality' in types:
                    address_components['sublocality'] = long_name
            
            return {
                'name': result.get('name', ''),
                'formatted_address': result.get('formatted_address', ''),
                'vicinity': result.get('vicinity', ''),
                'formatted_phone_number': result.get('formatted_phone_number', ''),
                'international_phone_number': result.get('international_phone_number', ''),
                'rating': result.get('rating'),
                'types': result.get('types', []),
                'business_status': result.get('business_status', ''),
                'website': result.get('website', ''),
                'address_components': address_components,
                'geometry': result.get('geometry', {}).get('location', {}),
                'opening_hours': result.get('opening_hours', {}).get('weekday_text', [])
            }
        else:
            return {'error': f"Place Details API Error: {data['status']}"}
            
    except Exception as e:
        return {'error': f"Exception in Place Details: {str(e)}"}

def search_place_text_search(row):
    """Use Google Places API Text Search to find a place and return place_id.
    Text Search is good for finding businesses by name and location."""
    
    query = build_place_search_query(row)
    if not query or query.isspace():
        return None, None, None, None, "Empty query", None
    
    # Build location bias if we have city/state
    location_bias = ""
    if pd.notna(row.get('city')) and pd.notna(row.get('state')):
        city = str(row['city']).strip()
        state = str(row['state']).strip()
        location_bias = f"&locationbias=point:{city},{state}"
    
    url = f'https://maps.googleapis.com/maps/api/place/textsearch/json?query={quote(query)}&region=us{location_bias}&key={API_KEY}'
    
    try:
        response = requests.get(url)
        data = response.json()
        
        if data['status'] == 'OK' and len(data['results']) > 0:
            result = data['results'][0]  # Take the first/best match
            
            place_id = result.get('place_id', '')
            name = result.get('name', '')
            formatted_address = result.get('formatted_address', '')
            vicinity = result.get('vicinity', '')
            
            # Get coordinates
            location = result.get('geometry', {}).get('location', {})
            lat = location.get('lat')
            lng = location.get('lng')
            
            # Get place types
            place_types = result.get('types', [])
            
            # Get basic info
            rating = result.get('rating')
            business_status = result.get('business_status', 'OPERATIONAL')
            
            # Get detailed information using Place Details API
            details = get_place_details(place_id) if place_id else {}
            
            return {
                'place_id': place_id,
                'name': details.get('name', name),
                'formatted_address': details.get('formatted_address', formatted_address),
                'vicinity': details.get('vicinity', vicinity),
                'latitude': lat,
                'longitude': lng,
                'rating': details.get('rating', rating),
                'place_types': details.get('types', place_types),
                'business_status': details.get('business_status', business_status),
                'formatted_phone_number': details.get('formatted_phone_number', ''),
                'international_phone_number': details.get('international_phone_number', ''),
                'website': details.get('website', ''),
                'address_components': details.get('address_components', {}),
                'opening_hours': details.get('opening_hours', []),
                'status': 'SUCCESS'
            }
        else:
            return {'status': f"API Error: {data['status']}"}
    except Exception as e:
        return {'status': f"Exception: {str(e)}"}

def search_place_nearby_search(row, radius=2000, expected_place_types=None):
    """Use Google Places API Nearby Search to find a place.
    Uses ZIP code for more precise location when available, falls back to city/state.
    Smaller radius for more targeted results.
    
    Args:
        row: DataFrame row with business information
        radius: Search radius in meters (default 2000)
        expected_place_types: List/string of expected place types for this business
    """
    
    # We need at least city/state or coordinates for nearby search
    if not (pd.notna(row.get('city')) and pd.notna(row.get('state'))):
        return {'status': "Missing location data for nearby search"}
    
    # Build the most precise location string possible
    # Priority: ZIP code > city/state
    if pd.notna(row.get('zip_code')):
        location_string = f"{row['zip_code']}, {row['state']}"
        print(f"    Using ZIP code for location: {location_string}")
    else:
        location_string = f"{row['city']}, {row['state']}"
        print(f"    Using city/state for location: {location_string}")
    
    # Get coordinates for the location
    geocode_url = f'https://maps.googleapis.com/maps/api/geocode/json?address={quote(location_string)}&key={API_KEY}'
    
    try:
        geocode_response = requests.get(geocode_url)
        geocode_data = geocode_response.json()
        
        if geocode_data['status'] != 'OK':
            return {'status': f"Geocoding failed: {geocode_data['status']}"}
        
        location = geocode_data['results'][0]['geometry']['location']
        lat = location['lat']
        lng = location['lng']
        
        # Build comprehensive keyword from business name (both chain and label)
        keyword_parts = []
        if pd.notna(row.get('chain')):
            keyword_parts.append(str(row['chain']).strip())
        if pd.notna(row.get('label')):
            # Only add label if it's different from chain to avoid redundancy
            label = str(row['label']).strip()
            chain = str(row.get('chain', '')).strip()
            if label and label.lower() != chain.lower():
                keyword_parts.append(label)
        
        keyword = ' '.join(keyword_parts) if keyword_parts else None
        print(f"    Using keyword: {keyword}")
        
        # Parse expected place types if provided
        expected_types = []
        if expected_place_types:
            if isinstance(expected_place_types, str):
                # Handle comma-separated string like your data
                expected_types = [t.strip() for t in expected_place_types.split(',') if t.strip()]
            elif isinstance(expected_place_types, list):
                expected_types = expected_place_types
        
        print(f"    Expected place types: {expected_types[:5]}..." if len(expected_types) > 5 else f"    Expected place types: {expected_types}")
        
        # Determine the best business type for API filtering
        # Priority: Use expected types > infer from keyword > default
        business_type = 'establishment'  # default
        
        if expected_types:
            # Use the most specific type from expected types
            type_priority = {
                'gas_station': 10,
                'convenience_store': 9, 
                'restaurant': 8,
                'rv_park': 7,
                'grocery_or_supermarket': 6,
                'car_wash': 5,
                'car_repair': 4,
                'atm': 3,
                'store': 2,
                'establishment': 1
            }
            
            best_type = None
            best_priority = 0
            for exp_type in expected_types:
                if exp_type in type_priority and type_priority[exp_type] > best_priority:
                    best_type = exp_type
                    best_priority = type_priority[exp_type]
            
            if best_type:
                business_type = best_type
                print(f"    Using expected type for API filter: {business_type}")
        
        # Fallback to keyword-based inference if no good expected type
        if business_type == 'establishment' and keyword:
            keyword_lower = keyword.lower()
            if any(food_term in keyword_lower for food_term in ['mcdonald', 'burger', 'pizza', 'restaurant', 'cafe', 'coffee', 'starbucks', 'subway', 'kfc', 'taco']):
                business_type = 'restaurant'
            elif any(gas_term in keyword_lower for gas_term in ['shell', 'exxon', 'bp', 'chevron', 'mobil', 'gas', 'station']):
                business_type = 'gas_station'
            elif any(store_term in keyword_lower for store_term in ['walmart', 'target', 'store', 'market', 'shop']):
                business_type = 'store'
            print(f"    Inferred type from keyword: {business_type}")
        
        print(f"    Final business type for API: {business_type}")
        
        # Build nearby search URL with smaller radius for more precision
        url = f'https://maps.googleapis.com/maps/api/place/nearbysearch/json?location={lat},{lng}&radius={radius}&key={API_KEY}'
        
        if keyword:
            url += f'&keyword={quote(keyword)}'
        
        # Add business type filter for better results
        url += f'&type={business_type}'
        
        response = requests.get(url)
        data = response.json()
        
        if data['status'] == 'OK' and len(data['results']) > 0:
            print(f"    Found {len(data['results'])} nearby results")
            
            # Enhanced matching algorithm with better scoring
            best_match = None
            best_score = 0
            
            # Build comprehensive search terms for matching
            search_terms = set()
            if keyword:
                # Add individual words from keyword
                search_terms.update(word.lower().strip() for word in keyword.split() if len(word.strip()) > 2)
            
            # Add address terms if available
            if pd.notna(row.get('address')):
                address_words = str(row['address']).lower().split()
                search_terms.update(word.strip() for word in address_words if len(word.strip()) > 2)
            
            print(f"    Search terms: {search_terms}")
            
            for i, result in enumerate(data['results']):
                result_name = result.get('name', '').lower()
                result_address = result.get('vicinity', '').lower()
                result_types = [t.lower() for t in result.get('types', [])]
                
                # Enhanced scoring algorithm with place type matching
                score = 0
                
                # Place type matching (highest weight if we have expected types)
                if expected_types:
                    type_matches = 0
                    for exp_type in expected_types:
                        if exp_type.lower() in result_types:
                            type_matches += 1
                    
                    if type_matches > 0:
                        # Heavy bonus for type matches - this is very important
                        type_score = type_matches * 15  # 15 points per matching type
                        score += type_score
                        print(f"      Result {i}: {type_matches} type matches (+{type_score})")
                    else:
                        # Penalty for no type matches when we have expectations
                        score -= 5
                        print(f"      Result {i}: No expected type matches (-5)")
                
                # Name matching (high weight)
                name_matches = 0
                for term in search_terms:
                    if term in result_name:
                        name_matches += 1
                        score += 10  # High weight for name matches
                
                if name_matches > 0:
                    print(f"      Result {i}: {name_matches} name matches (+{name_matches * 10})")
                
                # Address matching (medium weight)
                address_matches = 0
                for term in search_terms:
                    if term in result_address:
                        address_matches += 1
                        score += 3  # Medium weight for address matches
                
                if address_matches > 0:
                    print(f"      Result {i}: {address_matches} address matches (+{address_matches * 3})")
                
                # General business type matching (low weight) - only if no expected types
                if not expected_types and business_type.replace('_', ' ') in ' '.join(result_types):
                    score += 2
                    print(f"      Result {i}: General type match '{business_type}' (+2)")
                
                # Proximity bonus (very small) - all results get this
                score += 1  # Base score for being in search results
                
                print(f"      Result {i}: '{result.get('name')}' at '{result.get('vicinity')}' - Total Score: {score}")
                print(f"        Types: {result.get('types', [])}")
                
                if score > best_score:
                    best_score = score
                    best_match = result
            
            if not best_match:
                best_match = data['results'][0]  # Fallback to first result
                print(f"    No good matches found, using first result: {best_match.get('name')}")
            else:
                print(f"    Best match: {best_match.get('name')} (Score: {best_score})")
            
            place_id = best_match.get('place_id', '')
            name = best_match.get('name', '')
            formatted_address = best_match.get('vicinity', '')
            vicinity = best_match.get('vicinity', '')
            location = best_match.get('geometry', {}).get('location', {})
            result_lat = location.get('lat')
            result_lng = location.get('lng')
            rating = best_match.get('rating')
            place_types = best_match.get('types', [])
            business_status = best_match.get('business_status', 'OPERATIONAL')
            
            # Get detailed information using Place Details API
            details = get_place_details(place_id) if place_id else {}
            
            return {
                'place_id': place_id,
                'name': details.get('name', name),
                'formatted_address': details.get('formatted_address', formatted_address),
                'vicinity': details.get('vicinity', vicinity),
                'latitude': result_lat,
                'longitude': result_lng,
                'rating': details.get('rating', rating),
                'place_types': details.get('types', place_types),
                'business_status': details.get('business_status', business_status),
                'formatted_phone_number': details.get('formatted_phone_number', ''),
                'international_phone_number': details.get('international_phone_number', ''),
                'website': details.get('website', ''),
                'address_components': details.get('address_components', {}),
                'opening_hours': details.get('opening_hours', []),
                'status': 'SUCCESS'
            }
        else:
            return {'status': f"API Error: {data['status']}"}
            
    except Exception as e:
        return {'status': f"Exception: {str(e)}"}

def find_place_comprehensive(row, expected_place_types=None):
    """Try multiple Places API methods to find the best match for a business.
    Returns the best result from Text Search or Nearby Search.
    
    Args:
        row: DataFrame row with business information
        expected_place_types: List/string of expected place types for this business
    """
    
    # Try Text Search first (usually more accurate for businesses)
    result = search_place_text_search(row)
    
    if result.get('status') == "SUCCESS" and result.get('place_id'):
        result['search_method'] = 'TEXT_SEARCH'
        return result
    
    # If Text Search failed, try Nearby Search with expected types
    print(f"Text search failed for {row.get('chain', '')} {row.get('label', '')}, trying nearby search...")
    result = search_place_nearby_search(row, expected_place_types=expected_place_types)
    
    if result.get('status') == "SUCCESS" and result.get('place_id'):
        result['search_method'] = 'NEARBY_SEARCH'
        return result
    
    # Both methods failed - return empty result with failure status
    return {
        'place_id': None,
        'name': None,
        'formatted_address': None,
        'vicinity': None,
        'latitude': None,
        'longitude': None,
        'rating': None,
        'place_types': None,
        'business_status': None,
        'formatted_phone_number': None,
        'international_phone_number': None,
        'website': None,
        'address_components': {},
        'opening_hours': [],
        'search_method': 'FAILED',
        'status': f'FAILED: {result.get("status", "Unknown error")}'
    }
```

```python
# Simplified function focused on getting place_id and detailed information
def get_place_id_simple(chain=None, label=None, address=None, city=None, state=None, zip_code=None):
    """Simple function to get place_id and detailed info using Google Places API Text Search.
    
    Args:
        chain: Business chain name (e.g., 'McDonald\'s')
        label: Specific location label or name
        address: Street address
        city: City name
        state: State name
        zip_code: ZIP code
    
    Returns:
        dict: Complete place information including phone numbers, address components, etc.
    """
    
    # Build search query
    query_parts = []
    if chain: query_parts.append(str(chain))
    if label: query_parts.append(str(label))
    if address: query_parts.append(str(address))
    if city: query_parts.append(str(city))
    if state: query_parts.append(str(state))
    if zip_code: query_parts.append(str(zip_code))
    
    if not query_parts:
        return {'place_id': None, 'name': None, 'status': 'No search terms provided'}
    
    query = ', '.join(query_parts)
    
    # Make API request to Text Search
    url = f'https://maps.googleapis.com/maps/api/place/textsearch/json?query={quote(query)}&region=us&key={API_KEY}'
    
    try:
        response = requests.get(url)
        data = response.json()
        
        if data['status'] == 'OK' and len(data['results']) > 0:
            result = data['results'][0]
            place_id = result.get('place_id')
            
            if place_id:
                # Get detailed information using Place Details API
                details = get_place_details(place_id)
                
                if 'error' not in details:
                    return {
                        'place_id': place_id,
                        'name': details.get('name', ''),
                        'formatted_address': details.get('formatted_address', ''),
                        'vicinity': details.get('vicinity', ''),
                        'formatted_phone_number': details.get('formatted_phone_number', ''),
                        'international_phone_number': details.get('international_phone_number', ''),
                        'website': details.get('website', ''),
                        'rating': details.get('rating'),
                        'business_status': details.get('business_status', ''),
                        'place_types': details.get('types', []),
                        'address_components': details.get('address_components', {}),
                        'opening_hours': details.get('opening_hours', []),
                        'latitude': details.get('geometry', {}).get('lat'),
                        'longitude': details.get('geometry', {}).get('lng'),
                        'status': 'SUCCESS'
                    }
                else:
                    return {
                        'place_id': place_id,
                        'name': result.get('name', ''),
                        'status': f'Place Details failed: {details.get("error", "Unknown error")}'
                    }
            else:
                return {
                    'place_id': None,
                    'name': result.get('name', ''),
                    'status': 'No place_id returned'
                }
        else:
            return {
                'place_id': None,
                'name': None,
                'status': f'API Error: {data["status"]}'
            }
    except Exception as e:
        return {
            'place_id': None,
            'name': None,
            'status': f'Exception: {str(e)}'
        }

# Example usage of the enhanced simple function
print("Testing enhanced place_id lookup with detailed information:")

# Example 1: Search by business name and location
example1 = get_place_id_simple(
    chain="McDonald's",
    city="New York",
    state="NY"
)
print(f"Example 1 - McDonald's in NYC:")
print(f"  Place ID: {example1.get('place_id')}")
print(f"  Name: {example1.get('name')}")
print(f"  Phone: {example1.get('formatted_phone_number')}")
print(f"  Address: {example1.get('formatted_address')}")
print(f"  Status: {example1.get('status')}")

# Example 2: Search with full address
example2 = get_place_id_simple(
    chain="Starbucks",
    address="1234 Main St",
    city="Los Angeles",
    state="CA"
)
print(f"\nExample 2 - Starbucks with address:")
print(f"  Place ID: {example2.get('place_id')}")
print(f"  Name: {example2.get('name')}")
print(f"  Phone: {example2.get('formatted_phone_number')}")
print(f"  Website: {example2.get('website')}")
print(f"  Status: {example2.get('status')}")

# Example 3: Apply to your dataframe
print("\nApplying to sample data:")
for idx, row in random_df_2016.head(1).iterrows():  # Just test with first row
    result = get_place_id_simple(
        chain=row.get('chain'),
        label=row.get('label'),
        address=row.get('address'),
        city=row.get('city'),
        state=row.get('state'),
        zip_code=row.get('zip_code')
    )
    print(f"Row {idx} - {row.get('chain', '')} {row.get('label', '')}:")
    print(f"  Place ID: {result.get('place_id')}")
    print(f"  Name: {result.get('name')}")
    print(f"  Formatted Phone: {result.get('formatted_phone_number')}")
    print(f"  International Phone: {result.get('international_phone_number')}")
    print(f"  Address: {result.get('formatted_address')}")
    print(f"  Vicinity: {result.get('vicinity')}")
    print(f"  Website: {result.get('website')}")
    print(f"  Rating: {result.get('rating')}")
    print(f"  Business Status: {result.get('business_status')}")
    
    # Show address components
    addr_components = result.get('address_components', {})
    if addr_components:
        print("  Address Components:")
        for key, value in addr_components.items():
            if value:
                print(f"    {key}: {value}")
    
    if result.get('place_id'):
        maps_url = f"https://www.google.com/maps/place/?q=place_id:{result['place_id']}"
        print(f"  Google Maps Link: {maps_url}")
    
    print(f"  Status: {result.get('status')}")

# Display the Places API results
random_df_2016_places
```
