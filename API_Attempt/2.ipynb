```python
import pandas as pd
import requests
import time
from urllib.parse import quote
# Query Google Maps
from dotenv import load_dotenv
import os

# Load environment variables from .env file
load_dotenv()
API_KEY = os.getenv("API_KEY")  # This will get your API key from the environment

if not API_KEY:
    raise ValueError("API_KEY not found. Please set it in your .env file.")
```

```python
def build_address_string(row):
    """Build an address string from relevant columns."""
    components = []
    
    # Add components if they exist and aren't empty/NaN
    # Adjust these field names to match your actual column names
    if pd.notna(row.get('chain')): components.append(str(row['chain']))
    if pd.notna(row.get('label')): components.append(str(row['label']))
    if pd.notna(row.get('address')): components.append(str(row['address']))
    if pd.notna(row.get('city')): components.append(str(row['city']))
    if pd.notna(row.get('state')): components.append(str(row['state']))
    if pd.notna(row.get('zip_code')): components.append(str(row['zip_code']))
    
    # Join all components with commas
    return ', '.join(components)

def build_components_filter(row):
    """Build component filters for more precise geocoding."""
    components = ['country:US']  # Always restrict to US
    
    # Add specific components if they exist
    if pd.notna(row.get('city')):
        components.append(f"locality:{quote(str(row['city']))}")
    
    if pd.notna(row.get('state')):
        components.append(f"administrative_area:{quote(str(row['state']))}")
    
    if pd.notna(row.get('zip_code')):
        components.append(f"postal_code:{quote(str(row['zip_code']))}")
    
    return '|'.join(components)

def geocode_address_with_components(address_str, row=None):
    """Query Google Maps API to get lat/long for an address with component filtering.
    Uses structured components for better accuracy."""
    if not address_str or address_str.isspace():
        return None, None, None, "Empty address", None, None
    
    encoded_address = quote(address_str)
    
    # Build URL with component filtering
    base_url = f'https://maps.googleapis.com/maps/api/geocode/json?address={encoded_address}&region=us'
    
    # Add component filtering if row data is provided
    if row is not None:
        components_filter = build_components_filter(row)
        url = f'{base_url}&components={components_filter}&key={API_KEY}'
    else:
        # Fallback to country restriction only
        url = f'{base_url}&components=country:US&key={API_KEY}'
    
    try:
        response = requests.get(url)
        data = response.json()
        
        if data['status'] == 'OK':
            result = data['results'][0]
            lat = result['geometry']['location']['lat']
            lng = result['geometry']['location']['lng']
            
            # Track if this was a fuzzy/partial match
            is_partial = result.get('partial_match', False)
            
            # Get precision level
            precision = result['geometry']['location_type']
            
            # Get formatted address from Google (their best interpretation)
            formatted_address = result.get('formatted_address', '')
            
            # Get place_id for creating links
            place_id = result.get('place_id', '')
            
            match_type = "EXACT" if not is_partial else "PARTIAL"
            return lat, lng, precision, match_type, formatted_address, place_id
        else:
            return None, None, None, f"API Error: {data['status']}", None, None
    except Exception as e:
        return None, None, None, f"Exception: {str(e)}", None, None

# Keep the original function for backward compatibility
def geocode_address(address_str):
    """Query Google Maps API to get lat/long for an address with fuzzy match details.
    Biased toward US results."""
    if not address_str or address_str.isspace():
        return None, None, None, "Empty address", None, None
    
    encoded_address = quote(address_str)
    # Add region biasing for US and components restriction for better US results
    url = f'https://maps.googleapis.com/maps/api/geocode/json?address={encoded_address}&region=us&components=country:US&key={API_KEY}'
    
    try:
        response = requests.get(url)
        data = response.json()
        
        if data['status'] == 'OK':
            result = data['results'][0]
            lat = result['geometry']['location']['lat']
            lng = result['geometry']['location']['lng']
            
            # Track if this was a fuzzy/partial match
            is_partial = result.get('partial_match', False)
            
            # Get precision level
            precision = result['geometry']['location_type']
            
            # Get formatted address from Google (their best interpretation)
            formatted_address = result.get('formatted_address', '')
            
            # Get place_id for creating links
            place_id = result.get('place_id', '')
            
            match_type = "EXACT" if not is_partial else "PARTIAL"
            return lat, lng, precision, match_type, formatted_address, place_id
        else:
            return None, None, None, f"API Error: {data['status']}", None, None
    except Exception as e:
        return None, None, None, f"Exception: {str(e)}", None, None

```

```python
# Build address string for each row in random_df_2016
random_df_2016['address_for_geocoding'] = random_df_2016.apply(build_address_string, axis=1)

# Add columns to track geocoding results
random_df_2016['latitude'] = None
random_df_2016['longitude'] = None
random_df_2016['precision_level'] = None
random_df_2016['match_type'] = None
random_df_2016['google_formatted_address'] = None
random_df_2016['place_id'] = None
random_df_2016['google_maps_link'] = None
random_df_2016['geocoding_status'] = None

# Geocode each row in random_df_2016
for idx, row in random_df_2016.iterrows():
    address = row['address_for_geocoding']
    lat, lng, precision, match_type, formatted_address, place_id = geocode_address(address)
    status = 'SUCCESS' if lat is not None and lng is not None else 'FAILED'
    
    # Create Google Maps link
    if place_id:
        google_link = f"https://www.google.com/maps/place/?q=place_id:{place_id}"
    elif lat and lng:
        google_link = f"https://www.google.com/maps/@{lat},{lng},15z"
    else:
        google_link = None
    
    random_df_2016.at[idx, 'latitude'] = lat
    random_df_2016.at[idx, 'longitude'] = lng
    random_df_2016.at[idx, 'precision_level'] = precision
    random_df_2016.at[idx, 'match_type'] = match_type
    random_df_2016.at[idx, 'google_formatted_address'] = formatted_address
    random_df_2016.at[idx, 'place_id'] = place_id
    random_df_2016.at[idx, 'google_maps_link'] = google_link
    random_df_2016.at[idx, 'geocoding_status'] = status
    time.sleep(0.2)  # Respect Google API rate limits
    if idx % 10 == 0:
        print(f"Processed {idx} records")

# Show summary of geocoding results
print(f"Geocoding complete: {random_df_2016['geocoding_status'].value_counts().to_dict()}")
random_df_2016[['address_for_geocoding', 'latitude', 'longitude', 'precision_level', 'match_type', 'google_formatted_address', 'place_id', 'google_maps_link', 'geocoding_status']].head()
```

```python
# Example of using component-filtered geocoding for better accuracy
# This version uses structured components (city->locality, state->administrative_area, zip->postal_code)

# Create a copy for testing component filtering
random_df_2016_components = random_df_2016.copy()

# Build address string for each row
random_df_2016_components['address_for_geocoding'] = random_df_2016_components.apply(build_address_string, axis=1)

# Add columns to track geocoding results with component filtering
random_df_2016_components['latitude_comp'] = None
random_df_2016_components['longitude_comp'] = None
random_df_2016_components['precision_level_comp'] = None
random_df_2016_components['match_type_comp'] = None
random_df_2016_components['google_formatted_address_comp'] = None
random_df_2016_components['place_id_comp'] = None
random_df_2016_components['google_maps_link_comp'] = None
random_df_2016_components['geocoding_status_comp'] = None

# Geocode each row using component filtering
for idx, row in random_df_2016_components.iterrows():
    address = row['address_for_geocoding']
    lat, lng, precision, match_type, formatted_address, place_id = geocode_address_with_components(address, row)
    status = 'SUCCESS' if lat is not None and lng is not None else 'FAILED'
    
    # Create Google Maps link
    if place_id:
        google_link = f"https://www.google.com/maps/place/?q=place_id:{place_id}"
    elif lat and lng:
        google_link = f"https://www.google.com/maps/@{lat},{lng},15z"
    else:
        google_link = None
    
    random_df_2016_components.at[idx, 'latitude_comp'] = lat
    random_df_2016_components.at[idx, 'longitude_comp'] = lng
    random_df_2016_components.at[idx, 'precision_level_comp'] = precision
    random_df_2016_components.at[idx, 'match_type_comp'] = match_type
    random_df_2016_components.at[idx, 'google_formatted_address_comp'] = formatted_address
    random_df_2016_components.at[idx, 'place_id_comp'] = place_id
    random_df_2016_components.at[idx, 'google_maps_link_comp'] = google_link
    random_df_2016_components.at[idx, 'geocoding_status_comp'] = status
    time.sleep(0.2)  # Respect Google API rate limits
    if idx % 10 == 0:
        print(f"Processed {idx} records with component filtering")

# Show summary of component-filtered geocoding results
print(f"Component-filtered geocoding complete: {random_df_2016_components['geocoding_status_comp'].value_counts().to_dict()}")
random_df_2016_components[['address_for_geocoding', 'city', 'state', 'zip_code', 'latitude_comp', 'longitude_comp', 'precision_level_comp', 'match_type_comp', 'google_formatted_address_comp']].head()
```
