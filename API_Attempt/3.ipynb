```python
import pandas as pd
#import plt


df = pd.read_csv(r'C:\Users\clint\Desktop\Geocoding_Task\2.csv')
df
```

```python
# Comprehensive Analysis of precision_level column

print("=== PRECISION LEVEL ANALYSIS ===")
print(f"\nDataset shape: {df.shape}")
print(f"Total records: {len(df)}")

# Basic statistics for precision_level
print("\n1. BASIC STATISTICS:")
print(f"Unique precision levels: {df['precision_level'].nunique()}")
print(f"Unique values: {df['precision_level'].unique().tolist()}")
print(f"Missing values: {df['precision_level'].isnull().sum()}")

# Value counts and percentages
print("\n2. DISTRIBUTION:")
precision_counts = df['precision_level'].value_counts()
precision_pct = df['precision_level'].value_counts(normalize=True) * 100

for level in precision_counts.index:
    count = precision_counts[level]
    pct = precision_pct[level]
    print(f"{level}: {count} records ({pct:.1f}%)")

# Display the value counts as a DataFrame for better formatting
print("\n3. SUMMARY TABLE:")
summary_df = pd.DataFrame({
    'Count': precision_counts,
    'Percentage': precision_pct.round(1)
})
print(summary_df)
```

```python
# Visualizations for precision_level
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
plt.figure(figsize=(15, 12))

# 1. Bar chart of counts
plt.subplot(2, 2, 1)
sns.barplot(x=precision_counts.values, y=precision_counts.index, palette='Set2')
plt.title('Count of Records by Precision Level', fontsize=14, fontweight='bold')
plt.xlabel('Number of Records', fontsize=12)
plt.ylabel('Precision Level', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Add count labels on bars
for i, v in enumerate(precision_counts.values):
    plt.text(v + 2, i, str(v), va='center', fontweight='bold')

# 2. Pie chart with percentages
plt.subplot(2, 2, 2)
colors = sns.color_palette('Set2', len(precision_counts))
wedges, texts, autotexts = plt.pie(precision_counts.values, 
                                  labels=precision_counts.index,
                                  autopct='%1.1f%%',
                                  startangle=90,
                                  colors=colors,
                                  explode=[0.05]*len(precision_counts))
plt.title('Distribution of Precision Levels', fontsize=14, fontweight='bold')

# Enhance pie chart text
for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontweight('bold')

# 3. Horizontal bar chart with percentages
plt.subplot(2, 2, 3)
sns.barplot(x=precision_pct.values, y=precision_pct.index, palette='viridis')
plt.title('Percentage Distribution of Precision Levels', fontsize=14, fontweight='bold')
plt.xlabel('Percentage (%)', fontsize=12)
plt.ylabel('Precision Level', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Add percentage labels
for i, v in enumerate(precision_pct.values):
    plt.text(v + 0.5, i, f'{v:.1f}%', va='center', fontweight='bold')

# 4. Combined view with both counts and percentages
plt.subplot(2, 2, 4)
x_pos = np.arange(len(precision_counts))

# Create bars
bars = plt.bar(x_pos, precision_counts.values, color=colors, alpha=0.8, edgecolor='black')

# Customize the plot
plt.title('Precision Level Summary', fontsize=14, fontweight='bold')
plt.xlabel('Precision Level', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(x_pos, precision_counts.index, rotation=45, ha='right')

# Add labels on top of bars (count and percentage)
for i, (bar, count, pct) in enumerate(zip(bars, precision_counts.values, precision_pct.values)):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 2,
             f'{count}\n({pct:.1f}%)', ha='center', va='bottom', fontweight='bold', fontsize=10)

plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout(pad=3)
plt.show()
```

```python
# Quality Assessment based on Precision Levels
# Geocoding precision levels ranked by accuracy (best to worst)

print("=== PRECISION LEVEL QUALITY ASSESSMENT ===")

# Define precision level hierarchy (from most to least accurate)
precision_hierarchy = {
    'ROOFTOP': 'Highest - Exact building location',
    'RANGE_INTERPOLATED': 'High - Interpolated along street segment', 
    'GEOMETRIC_CENTER': 'Medium - Center of address range/area',
    'APPROXIMATE': 'Lowest - General area approximation'
}

print("\nPRECISION LEVEL DEFINITIONS (by accuracy):")
for level, description in precision_hierarchy.items():
    if level in precision_counts.index:
        count = precision_counts[level]
        pct = precision_pct[level]
        print(f" {level}: {description}")
        print(f"   {count} records ({pct:.1f}% of dataset)")
    else:
        print(f" {level}: {description}")
        print(f"   0 records (0.0% of dataset)")

# Calculate quality tiers
high_quality = ['ROOFTOP']
medium_quality = ['RANGE_INTERPOLATED', 'GEOMETRIC_CENTER'] 
low_quality = ['APPROXIMATE']

high_count = sum(precision_counts.get(level, 0) for level in high_quality)
medium_count = sum(precision_counts.get(level, 0) for level in medium_quality)
low_count = sum(precision_counts.get(level, 0) for level in low_quality)

high_pct = (high_count / len(df)) * 100
medium_pct = (medium_count / len(df)) * 100  
low_pct = (low_count / len(df)) * 100

print("\n=== QUALITY TIER SUMMARY ===")
print(f"High Quality (ROOFTOP): {high_count} records ({high_pct:.1f}%)")
print(f"Medium Quality (RANGE_INTERPOLATED, GEOMETRIC_CENTER): {medium_count} records ({medium_pct:.1f}%)")
print(f"Low Quality (APPROXIMATE): {low_count} records ({low_pct:.1f}%)")

# Overall quality score (weighted average)
quality_weights = {'ROOFTOP': 4, 'RANGE_INTERPOLATED': 3, 'GEOMETRIC_CENTER': 2, 'APPROXIMATE': 1}
total_weighted_score = sum(precision_counts.get(level, 0) * weight for level, weight in quality_weights.items())
max_possible_score = len(df) * 4
overall_quality_score = (total_weighted_score / max_possible_score) * 100

print(f"\nOVERALL QUALITY SCORE: {overall_quality_score:.1f}% (out of 100%)")
print("(Based on weighted average: ROOFTOP=4, RANGE_INTERPOLATED=3, GEOMETRIC_CENTER=2, APPROXIMATE=1)")
```

## Findings from match_type Analysis

1. **Distribution**: The data is heavily skewed towards 'PARTIAL' match types:
   - PARTIAL: 98.2% of all records (383 entries)
   - EXACT: Only 1.8% of all records (7 entries)

2. **Relationship with precision_level**:
   - All EXACT matches (100%) have a ROOFTOP precision level, which indicates high accuracy geocoding
   - PARTIAL matches are distributed across different precision levels:
     - ROOFTOP: 46.5% 
     - GEOMETRIC_CENTER: 25.8%
     - RANGE_INTERPOLATED: 14.9%
     - APPROXIMATE: 12.8%

3. **Implications**:
   - The dataset primarily consists of partially matched addresses
   - Even with partial matches, a significant portion (46.5%) still achieves ROOFTOP precision
   - The low number of EXACT matches suggests potential issues with address formatting or discrepancies between source addresses and reference database

```python
# Enhanced Analysis: High-Quality vs Lower-Quality Geocoding Results
# High-Quality = ROOFTOP precision_level AND/OR EXACT match_type
# Lower-Quality = Everything else

import numpy as np
from scipy import stats

# Create quality categories
# High-Quality: ROOFTOP precision OR EXACT match
df['quality_category'] = 'Lower-Quality'
df.loc[(df['precision_level'] == 'ROOFTOP') | (df['match_type'] == 'EXACT'), 'quality_category'] = 'High-Quality'

# Alternative stricter definition: BOTH ROOFTOP AND EXACT
df['strict_quality_category'] = 'Lower-Quality'
df.loc[(df['precision_level'] == 'ROOFTOP') & (df['match_type'] == 'EXACT'), 'strict_quality_category'] = 'High-Quality (Strict)'

# Count and percentage analysis
quality_counts = df['quality_category'].value_counts()
quality_pct = df['quality_category'].value_counts(normalize=True) * 100

strict_quality_counts = df['strict_quality_category'].value_counts()
strict_quality_pct = df['strict_quality_category'].value_counts(normalize=True) * 100

print("=== QUALITY ANALYSIS RESULTS ===")
print("\n1. INCLUSIVE DEFINITION (ROOFTOP OR EXACT):")
print(f"High-Quality matches: {quality_counts.get('High-Quality', 0)} ({quality_pct.get('High-Quality', 0):.1f}%)")
print(f"Lower-Quality matches: {quality_counts.get('Lower-Quality', 0)} ({quality_pct.get('Lower-Quality', 0):.1f}%)")

print("\n2. STRICT DEFINITION (ROOFTOP AND EXACT):")
print(f"High-Quality matches: {strict_quality_counts.get('High-Quality (Strict)', 0)} ({strict_quality_pct.get('High-Quality (Strict)', 0):.1f}%)")
print(f"Lower-Quality matches: {strict_quality_counts.get('Lower-Quality', 0)} ({strict_quality_pct.get('Lower-Quality', 0):.1f}%)")

# Detailed breakdown
print("\n=== DETAILED BREAKDOWN ===")
breakdown = pd.crosstab(df['precision_level'], df['match_type'], margins=True)
print("\nCross-tabulation of precision_level vs match_type:")
print(breakdown)

# Calculate percentages for cross-tab
breakdown_pct = pd.crosstab(df['precision_level'], df['match_type'], normalize='all') * 100
print("\nPercentage distribution:")
print(breakdown_pct.round(1))
```
