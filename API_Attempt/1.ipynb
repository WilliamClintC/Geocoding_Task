```python
def standardize_highway_address(address):
    """
    Standardize highway address formats to be more consistent.
    
    Examples:
uld be    - '1-80 Exit 162 ( UT 280 )' -> 'I-80 Exit 162 and UT 280'
    - 'I - 80 Exit 162 ( UT 280 )' -> 'I-80 Exit 162 and UT 280'
    - '1-15-84 Exit 357' -> 'I-15 and I-84 Exit 357'
    - 'US 191 & US 491' -> 'US 191 and US 491'
    - '1-70 Exit 56 ( US 89 )' -> 'I-70 Exit 56 and US 89'
    - '1-15 Exit 62 ( N Interchange )' -> 'I-15 Exit 62 and N Interchange'
    """
    if not address or pd.isna(address):
        return address
    
    # Convert to string and strip whitespace
    addr = str(address).strip()
    
    # Step 1: Fix interstate numbering (1-XX -> I-XX)
    # Handle cases like "1-80", "1-15", etc. but not "US 191"
    import re
    
    # Replace standalone "1-" with "I-" at the beginning or after spaces
    addr = re.sub(r'\b1-(?=\d)', 'I-', addr)
    
    # Step 2: Remove extra spaces around dashes in interstate names
    # "I - 80" -> "I-80"
    addr = re.sub(r'I\s*-\s*(\d+)', r'I-\1', addr)
    
    # Step 3: Handle multiple interstates like "1-15-84" -> "I-15 and I-84"
    # Look for patterns like "I-XX-YY" and split them
    addr = re.sub(r'I-(\d+)-(\d+)', r'I-\1 and I-\2', addr)
    
    # Step 4: Replace "&" with "and"
    addr = re.sub(r'\s*&\s*', ' and ', addr)
    
    # Step 4.5: Replace "/" with "and" (but preserve fractions like 1/2, 3/4, etc.)
    # "15 X 305 B / I-80 WB X 122 / UT 201 EB X17" -> "15 X 305 B and I-80 WB X 122 and UT 201 EB X17"
    # But keep fractions: "1/2 mi S" should stay "1/2 mi S", not "1 and 2 mi S"
    
    # First, temporarily replace fractions with placeholders to protect them
    fraction_pattern = r'\b(\d+)/(\d+)\b'
    fractions = re.findall(fraction_pattern, addr)
    fraction_placeholders = {}
    
    for i, (num, denom) in enumerate(fractions):
        placeholder = f"__FRACTION_{i}__"
        fraction_placeholders[placeholder] = f"{num}/{denom}"
        addr = addr.replace(f"{num}/{denom}", placeholder, 1)
    
    # Now replace remaining forward slashes with "and"
    addr = re.sub(r'\s*/\s*', ' and ', addr)
    
    # Restore the fractions
    for placeholder, fraction in fraction_placeholders.items():
        addr = addr.replace(placeholder, fraction)
    
    # Step 5: Handle parentheses - replace with "and"
    # "Exit 162 ( UT 280 )" -> "Exit 162 and UT 280"
    # "Exit 62 ( N Interchange )" -> "Exit 62 and N Interchange"
    # Handle complete parentheses pairs first
    addr = re.sub(r'\s*\(\s*(.+?)\s*\)', r' and \1', addr)
    
    # Handle single opening parenthesis without closing
    # "Exit 85 ( W El" -> "Exit 85 and W El"
    addr = re.sub(r'\s*\(\s*(.+?)$', r' and \1', addr)
    
    # Step 6: Clean up multiple spaces
    addr = re.sub(r'\s+', ' ', addr)
    
    # Step 7: Remove trailing/leading whitespace
    addr = addr.strip()
    
    return addr

# Test the function with your examples
test_cases = [
    '1-80 Exit 162 ( UT 280 )',
    'I - 80 Exit 162 ( UT 280 )',
    '1-15-84 Exit 357',
    'US 191 & US 491',
    '1-70 Exit 56 ( US 89 )',
    '1-15 Exit 62 ( N Interchange )',
    '1-80 Exit 85 ( W El )',
    '1-80 Exit 85 ( W El',  # Single opening parenthesis case
    '1590 US 40 S ( 1/2 mi S of jct US 189 )',  # Test fraction handling
    'Exit 15 / I-84 Junction',  # Non-fraction forward slash
    'Highway 89 / State Route 12',  # Non-fraction forward slash
    'US 191/US 491',  # Non-fraction forward slash
    '3/4 mile west of I-15/I-84',  # Mixed: fraction + non-fraction slash
]

print("Testing highway address standardization:")
print("=" * 50)
for test in test_cases:
    result = standardize_highway_address(test)
    print(f"'{test}' -> '{result}'")
    
print("\n" + "=" * 50)
```

```python
# Apply the standardize_highway_address function to the address column
western_df['standardized_address'] = western_df['address'].apply(standardize_highway_address)

# Show before and after comparison for addresses that changed
changed_addresses = western_df[western_df['address'] != western_df['standardized_address']]

if len(changed_addresses) > 0:
    print(f"Found {len(changed_addresses)} addresses that were standardized:")
    print("=" * 80)
    for idx, row in changed_addresses.head(10).iterrows():  # Show first 10 changes
        print(f"Original:     '{row['address']}'")
        print(f"Standardized: '{row['standardized_address']}'")
        print("-" * 40)
else:
    print("No addresses needed standardization")

# Display summary
print(f"\nTotal addresses processed: {len(western_df)}")
print(f"Addresses changed: {len(changed_addresses)}")
```

```python
def replace_parentheses_with_commas(label):
    """
    Replace parentheses with commas in a label string.
    
    Examples:
    - 'Holiday Hills ( 66 )' -> 'Holiday Hills ,66,'
    - 'Store Name (Location)' -> 'Store Name ,Location,'
    - 'Normal Label' -> 'Normal Label' (unchanged)
    """
    if not label or pd.isna(label):
        return label
    
    # Convert to string and strip whitespace
    label_str = str(label).strip()
    
    # Replace opening parenthesis with comma
    label_str = label_str.replace('(', ',')
    
    # Replace closing parenthesis with comma
    label_str = label_str.replace(')', ',')
    
    # Clean up multiple spaces that might result from the replacement
    import re
    label_str = re.sub(r'\s+', ' ', label_str)
    
    # Clean up any double commas or spaces around commas
    label_str = re.sub(r'\s*,\s*', ',', label_str)
    
    return label_str

# Test the function with some examples
test_labels = [
    'Holiday Hills ( 66 )',
    'Store Name (Location)',
    'Normal Label',
    'Multiple (Parentheses) (Here)',
    None,
    'No Parentheses Here'
]

print("Testing parentheses to comma conversion:")
print("=" * 50)
for test in test_labels:
    result = replace_parentheses_with_commas(test)
    print(f"'{test}' -> '{result}'")

print("\n" + "=" * 50)
```

```python
# Apply the function to the label column
western_df['label'] = western_df['label'].apply(replace_parentheses_with_commas)

# Show labels that were changed
print("Sample of processed labels:")
print("=" * 40)
for idx, label in western_df['label'].head(10).items():
    print(f"Row {idx}: '{label}'")

print(f"\nTotal labels processed: {len(western_df)}")

# Check if any labels contain parentheses after processing
labels_with_parentheses = western_df[western_df['label'].str.contains(r'[()]', na=False, regex=True)]
print(f"Labels still containing parentheses: {len(labels_with_parentheses)}")

# Show the first few rows to verify the changes
print("\nFirst 5 rows with updated labels:")
print(western_df[['chain', 'label', 'address', 'city', 'state']].head())
```
