```python
# Check the columns in both dataframes
print("df columns:", df.columns.tolist())
print("\ndf2 columns:", df2.columns.tolist())

# Check if phone column exists in df
if 'phone' in df.columns:
    print(f"\ndf['phone'] sample values:")
    print(df['phone'].head())
    print(f"df['phone'] data type: {df['phone'].dtype}")
    print(f"df['phone'] non-null count: {df['phone'].notna().sum()}")
else:
    print("\n'phone' column not found in df")

# Check if Original_Phone column exists in df2
if 'Original_Phone' in df2.columns:
    print(f"\ndf2['Original_Phone'] sample values:")
    print(df2['Original_Phone'].head())
    print(f"df2['Original_Phone'] data type: {df2['Original_Phone'].dtype}")
    print(f"df2['Original_Phone'] non-null count: {df2['Original_Phone'].notna().sum()}")
else:
    print("\n'Original_Phone' column not found in df2")

# Check if STATUS column exists in df2
if 'STATUS' in df2.columns:
    print(f"\ndf2['STATUS'] sample values:")
    print(df2['STATUS'].head())
    print(f"df2['STATUS'] unique values: {df2['STATUS'].unique()}")
else:
    print("\n'STATUS' column not found in df2")
```

```python
# Phone number matching logic
import numpy as np

# First, let's identify the correct column names for phone numbers
phone_col_df = None
phone_col_df2 = None
status_col_df2 = None

# Check for phone column in df
for col in df.columns:
    if 'phone' in col.lower():
        phone_col_df = col
        break

# Check for phone column in df2
for col in df2.columns:
    if 'phone' in col.lower():
        phone_col_df2 = col
        break

# Check for status column in df2
for col in df2.columns:
    if 'status' in col.lower():
        status_col_df2 = col
        break

print(f"Using phone column from df: {phone_col_df}")
print(f"Using phone column from df2: {phone_col_df2}")
print(f"Using status column from df2: {status_col_df2}")

if phone_col_df is None or phone_col_df2 is None:
    print("ERROR: Could not find phone columns in one or both dataframes")
    print("df columns:", df.columns.tolist())
    print("df2 columns:", df2.columns.tolist())
else:
    # Initialize the new columns
    df['Phone_Yellowpages_matches_row_ids'] = None
    df['Yellowbook_phone_match_rate'] = True
    
    # Convert phone columns to string and handle NaN values
    df[phone_col_df] = df[phone_col_df].astype(str)
    df2[phone_col_df2] = df2[phone_col_df2].astype(str)
    
    # Create a dictionary for faster lookup from df2
    # Map phone numbers to their row indices
    phone_to_indices = {}
    for idx, phone in enumerate(df2[phone_col_df2]):
        if phone != 'nan' and phone != '':  # Skip NaN and empty values
            if phone not in phone_to_indices:
                phone_to_indices[phone] = []
            phone_to_indices[phone].append(idx)
    
    print(f"Created phone lookup dictionary with {len(phone_to_indices)} unique phone numbers")
    
    # Process each row in df
    for df_idx, row in df.iterrows():
        phone = row[phone_col_df]
        
        # Skip if phone is NaN or empty
        if phone == 'nan' or phone == '':
            df.at[df_idx, 'Phone_Yellowpages_matches_row_ids'] = []
            df.at[df_idx, 'Yellowbook_phone_match_rate'] = True
            continue
        
        # Find matching indices in df2
        if phone in phone_to_indices:
            matching_indices = phone_to_indices[phone]
            df.at[df_idx, 'Phone_Yellowpages_matches_row_ids'] = matching_indices
            
            # Check if any of the matching rows have STATUS == "no_results_found"
            match_rate = True
            if status_col_df2 is not None:
                for match_idx in matching_indices:
                    if df2.iloc[match_idx][status_col_df2] == 'no_results_found':
                        match_rate = False
                        break
            
            df.at[df_idx, 'Yellowbook_phone_match_rate'] = match_rate
        else:
            # No matches found
            df.at[df_idx, 'Phone_Yellowpages_matches_row_ids'] = []
            df.at[df_idx, 'Yellowbook_phone_match_rate'] = True
    
    print("Phone matching completed!")
    print(f"Total rows processed: {len(df)}")
    print(f"Rows with matches: {df['Phone_Yellowpages_matches_row_ids'].apply(lambda x: len(x) > 0 if isinstance(x, list) else False).sum()}")
    print(f"Rows with Yellowbook_phone_match_rate = False: {(df['Yellowbook_phone_match_rate'] == False).sum()}")
```

```python
# Check all columns in df2 that might contain phone numbers
print("=== All columns in df2 ===")
for i, col in enumerate(df2.columns):
    print(f"{i}: {col}")

# Look for columns that might have phone numbers in the format 435-336-4421
print("\n=== Looking for phone-like patterns ===")
for col in df2.columns:
    if df2[col].dtype == 'object':  # Only check string columns
        # Check if any values match phone format pattern
        sample_values = df2[col].dropna().head(10).astype(str)
        phone_like = [val for val in sample_values if '-' in val and len(val) == 12]
        if phone_like:
            print(f"Column '{col}' has phone-like values: {phone_like[:3]}")

# Specifically look for Original_Phone or similar
original_phone_cols = [col for col in df2.columns if 'original' in col.lower() or 'phone' in col.lower()]
print(f"\nColumns with 'original' or 'phone': {original_phone_cols}")

# Check the first few rows of df2 to understand the structure
print("\n=== First 2 rows of df2 ===")
print(df2.head(2).T)  # Transpose to see all columns
```

```python
# Based on the output, let's create the corrected matching logic
# It seems like we need to check which column actually contains the phone numbers we're looking for

# First, let's check if there's a column that might be the original phone number
# Check for any column that might contain the phone format we're looking for
phone_col_df2 = None

# Check each column for phone number patterns
for col in df2.columns:
    if df2[col].dtype == 'object':
        # Check if this column contains phone numbers in the format XXX-XXX-XXXX
        sample_non_null = df2[col].dropna()
        if len(sample_non_null) > 0:
            # Convert to string and check format
            sample_str = sample_non_null.astype(str).head(100)
            phone_pattern_count = sum(1 for x in sample_str if len(x) == 12 and x.count('-') == 2)
            
            if phone_pattern_count > 0:
                print(f"Column '{col}' contains {phone_pattern_count} phone-like values out of {len(sample_str)} sampled")
                # If this looks like a phone column, use it
                if phone_col_df2 is None or phone_pattern_count > 50:  # Use the column with most phone-like values
                    phone_col_df2 = col

print(f"Selected phone column from df2: {phone_col_df2}")

# Now let's redo the matching with the correct column
if phone_col_df2 is not None:
    # Initialize the new columns (reset them)
    df['Phone_Yellowpages_matches_row_ids'] = None
    df['Yellowbook_phone_match_rate'] = True
    
    # Convert phone columns to string and handle NaN values
    df['phone'] = df['phone'].astype(str)
    df2[phone_col_df2] = df2[phone_col_df2].astype(str)
    
    # Create a dictionary for faster lookup from df2
    phone_to_indices = {}
    for idx, phone in enumerate(df2[phone_col_df2]):
        if phone != 'nan' and phone != '' and phone != 'None':
            if phone not in phone_to_indices:
                phone_to_indices[phone] = []
            phone_to_indices[phone].append(idx)
    
    print(f"Created phone lookup dictionary with {len(phone_to_indices)} unique phone numbers")
    
    # Show some sample phones from both dataframes for verification
    print("\nSample phones from df['phone']:")
    print(df['phone'].dropna().head(5).tolist())
    print(f"\nSample phones from df2['{phone_col_df2}']:")
    print(df2[phone_col_df2].dropna().head(5).tolist())
    
    # Process each row in df
    rows_with_matches = 0
    rows_with_valid_status = 0
    
    for df_idx, row in df.iterrows():
        phone = row['phone']
        
        # Skip if phone is NaN or empty
        if phone == 'nan' or phone == '' or phone == 'None':
            df.at[df_idx, 'Phone_Yellowpages_matches_row_ids'] = []
            df.at[df_idx, 'Yellowbook_phone_match_rate'] = True
            continue
        
        # Find matching indices in df2
        if phone in phone_to_indices:
            matching_indices = phone_to_indices[phone]
            df.at[df_idx, 'Phone_Yellowpages_matches_row_ids'] = matching_indices
            rows_with_matches += 1
            
            # Check if any of the matching rows have STATUS == "no_results_found"
            match_rate = True
            for match_idx in matching_indices:
                if df2.iloc[match_idx]['STATUS'] == 'no_results_found':
                    match_rate = False
                    rows_with_valid_status += 1
                    break
            
            df.at[df_idx, 'Yellowbook_phone_match_rate'] = match_rate
        else:
            # No matches found
            df.at[df_idx, 'Phone_Yellowpages_matches_row_ids'] = []
            df.at[df_idx, 'Yellowbook_phone_match_rate'] = True
    
    print(f"\nMatching completed!")
    print(f"Total rows processed: {len(df)}")
    print(f"Rows with matches: {rows_with_matches}")
    print(f"Rows with valid status: {rows_with_valid_status}")
    print(f"Rows with Yellowbook_phone_match_rate = False: {(df['Yellowbook_phone_match_rate'] == False).sum()}")
    
    # Show some examples of matches
    matched_rows = df[df['Phone_Yellowpages_matches_row_ids'].apply(lambda x: len(x) > 0 if isinstance(x, list) else False)]
    if len(matched_rows) > 0:
        print(f"\nSample matched rows (first 3):")
        sample_matched = matched_rows.head(3)[['phone', 'Phone_Yellowpages_matches_row_ids', 'Yellowbook_phone_match_rate']]
        print(sample_matched)
    else:
        print("\nNo matches found - checking for potential issues...")
        # Check for common phone format mismatches
        df_phones = set(df['phone'].dropna().astype(str))
        df2_phones = set(df2[phone_col_df2].dropna().astype(str))
        
        # Check for exact matches
        exact_matches = df_phones & df2_phones
        print(f"Exact matches between phone sets: {len(exact_matches)}")
        
        # Show some examples from each set
        print("Sample from df phones:", list(df_phones)[:5])
        print("Sample from df2 phones:", list(df2_phones)[:5])
        
else:
    print("ERROR: Could not find a suitable phone column in df2")
```

```python
# Final verification and summary
print("=== FINAL RESULTS SUMMARY ===")
print(f"Total rows in df: {len(df)}")
print(f"Rows with phone matches: {df['Phone_Yellowpages_matches_row_ids'].apply(lambda x: len(x) > 0 if isinstance(x, list) else False).sum()}")
print(f"Rows with Yellowbook_phone_match_rate = False: {(df['Yellowbook_phone_match_rate'] == False).sum()}")
print(f"Rows with Yellowbook_phone_match_rate = True: {(df['Yellowbook_phone_match_rate'] == True).sum()}")

# Show the new columns
print("\n=== NEW COLUMNS ADDED ===")
print("1. Phone_Yellowpages_matches_row_ids: Contains list of matching row indices from df2")
print("2. Yellowbook_phone_match_rate: Boolean indicating if match has valid status")

# Show a sample of the results
print("\n=== SAMPLE RESULTS ===")
sample_cols = ['phone', 'Phone_Yellowpages_matches_row_ids', 'Yellowbook_phone_match_rate']
sample_data = df[sample_cols].head(10)
print(sample_data)

# Show rows with matches (if any)
matched_rows = df[df['Phone_Yellowpages_matches_row_ids'].apply(lambda x: len(x) > 0 if isinstance(x, list) else False)]
if len(matched_rows) > 0:
    print(f"\n=== ROWS WITH MATCHES (first 5) ===")
    print(matched_rows[sample_cols].head(5))
    
    # Show what the matches look like in df2
    print(f"\n=== CORRESPONDING MATCHES IN DF2 ===")
    first_match = matched_rows.iloc[0]
    match_indices = first_match['Phone_Yellowpages_matches_row_ids']
    print(f"Phone: {first_match['phone']}")
    print(f"Matching indices in df2: {match_indices}")
    print("Corresponding rows in df2:")
    print(df2.iloc[match_indices][['PHONE', 'STATUS']].head())
else:
    print("\n=== NO MATCHES FOUND ===")
    print("This could be due to:")
    print("1. Different phone number formats")
    print("2. Missing phone numbers in either dataset")
    print("3. Different column names than expected")

print("\n=== TASK COMPLETED ===")
print("The following columns have been added to df:")
print("- Phone_Yellowpages_matches_row_ids: List of matching row indices from df2")
print("- Yellowbook_phone_match_rate: Boolean (False if any match has STATUS='no_results_found')")
```
