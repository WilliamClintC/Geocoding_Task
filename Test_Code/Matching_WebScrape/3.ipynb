```python
# Setup and initialization
import numpy as np
import time

# Find phone column in df1
phone_col = None
if 'Phone' in df1.columns:
    phone_col = 'Phone'
elif 'phone' in df1.columns:
    phone_col = 'phone'
else:
    phone_cols = [col for col in df1.columns if 'phone' in col.lower()]
    if phone_cols:
        phone_col = phone_cols[0]

# Setup processing variables
all_rows = df1.index.tolist()
total_rows = len(all_rows)
search_columns = [col for col in df2.columns if col.lower() in ['phone', 'phone 2', 'fax']]

print(f"Processing {total_rows} rows from df1 using phone column: {phone_col}")
print(f"Searching in df2 columns: {search_columns}")

start_time = time.time()
```

```python
# Optimized phone number matching
all_matches = {}

# Pre-process df2 columns for faster comparison
df2_processed = df2.copy()
for col in search_columns:
    df2_processed[col] = df2_processed[col].astype(str)

# Get unique phone numbers to reduce redundant processing
df1_phones = df1[df1[phone_col].notna()][phone_col].astype(str)
unique_phones = df1_phones.unique()
print(f"Processing {len(unique_phones)} unique phone numbers (from {len(df1_phones)} total rows)")

# Process unique phone numbers with vectorized operations
phone_to_matches = {}
for i, phone_str in enumerate(unique_phones):
    if i % max(1, len(unique_phones) // 10) == 0:  # Progress every 10%
        print(f"Progress: {(i+1)/len(unique_phones)*100:.0f}% ({i+1}/{len(unique_phones)})")
    
    phone_matches = {}
    all_df2_matches = set()
    
    for col in search_columns:
        exact_mask = df2_processed[col] == phone_str
        exact_indices = df2_processed[exact_mask].index.tolist()
        
        partial_mask = df2_processed[col].str.contains(phone_str, case=False, na=False)
        partial_indices = df2_processed[partial_mask].index.tolist()
        
        phone_matches[col] = {
            'exact': len(exact_indices),
            'partial': len(partial_indices),
            'exact_indices': exact_indices,
            'partial_indices': partial_indices
        }
        
        all_df2_matches.update(exact_indices)
        all_df2_matches.update(partial_indices)
    
    total_exact = sum([phone_matches[col]['exact'] for col in phone_matches])
    total_partial = sum([phone_matches[col]['partial'] for col in phone_matches])
    
    phone_to_matches[phone_str] = {
        'matches': phone_matches,
        'total_exact': total_exact,
        'total_partial': total_partial,
        'all_df2_indices': sorted(list(all_df2_matches))
    }

# Map results back to all df1 rows
for row_idx in all_rows:
    if phone_col and pd.notna(df1.loc[row_idx, phone_col]):
        phone_str = str(df1.loc[row_idx, phone_col])
        match_data = phone_to_matches[phone_str]
        all_matches[row_idx] = {
            'phone_number': df1.loc[row_idx, phone_col],
            'matches': match_data['matches'],
            'total_exact': match_data['total_exact'],
            'total_partial': match_data['total_partial'],
            'all_df2_indices': match_data['all_df2_indices']
        }
    else:
        all_matches[row_idx] = {
            'phone_number': None,
            'matches': {},
            'total_exact': 0,
            'total_partial': 0,
            'all_df2_indices': []
        }

# Summary
total_df1_with_matches = sum(1 for match in all_matches.values() if match['all_df2_indices'])
total_unique_df2_matches = len(set().union(*[match['all_df2_indices'] for match in all_matches.values()]))
elapsed_time = time.time() - start_time

print(f"\nCompleted in {elapsed_time:.1f} seconds ({elapsed_time/60:.1f} minutes)")
print(f"df1 rows with matches: {total_df1_with_matches}/{len(all_matches)}")
print(f"Unique df2 rows involved: {total_unique_df2_matches}")
```

```python
# Create combined DataFrame with match accuracy information
print("Creating combined DataFrame...")

all_combined_rows = []

# Prepare renamed DataFrames
df1_full_renamed = df1.copy()
df1_full_renamed.columns = ['OCR_' + col for col in df1_full_renamed.columns]

df2_renamed = df2.copy()
df2_renamed.columns = ['Scraped_' + col for col in df2_renamed.columns]

# Process each row to create combined results
for i, row_idx in enumerate(all_rows):
    if i % max(1, total_rows // 10) == 0:  # Progress every 10%
        print(f"Progress: {(i+1)/total_rows*100:.0f}% ({i+1}/{total_rows})")
    
    match_data = all_matches[row_idx]
    
    # Get all unique matching df2 indices for this row
    match_accuracy_map = {}
    if match_data['matches']:
        for col, results in match_data['matches'].items():
            for df2_idx in results['exact_indices']:
                if df2_idx not in match_accuracy_map:
                    match_accuracy_map[df2_idx] = []
                match_accuracy_map[df2_idx].append(f"exact phone number ({col})")
            
            for df2_idx in results['partial_indices']:
                if df2_idx not in match_accuracy_map:
                    match_accuracy_map[df2_idx] = []
                if df2_idx not in results['exact_indices']:
                    match_accuracy_map[df2_idx].append(f"partial phone number ({col})")
    
    total_matches = len(match_accuracy_map)
    
    # Determine flagged status
    flag_reason = None
    if total_matches == 0:
        flag_reason = "No matches found"
    elif total_matches > 1:
        flag_reason = "Multiple matches found"
    
    flagged = flag_reason is not None
    
    # Create combined rows
    if total_matches > 0:
        for df2_idx, match_types in sorted(match_accuracy_map.items()):
            df1_row = df1_full_renamed.loc[row_idx].to_dict()
            df2_row = df2_renamed.iloc[df2_idx].to_dict()
            
            combined_row = {**df1_row, **df2_row}
            combined_row['match_accuracy'] = ', '.join(match_types)
            combined_row['df1_source_row'] = row_idx
            combined_row['df2_matched_row'] = df2_idx
            combined_row['Flagged'] = flagged
            combined_row['Flag_Reason'] = flag_reason
            
            all_combined_rows.append(combined_row)
    else:
        # No matches
        df1_row = df1_full_renamed.loc[row_idx].to_dict()
        empty_df2_row = {col: None for col in df2_renamed.columns}
        
        combined_row = {**df1_row, **empty_df2_row}
        combined_row['match_accuracy'] = "No matches"
        combined_row['df1_source_row'] = row_idx
        combined_row['df2_matched_row'] = None
        combined_row['Flagged'] = flagged
        combined_row['Flag_Reason'] = flag_reason
        
        all_combined_rows.append(combined_row)

combined_df = pd.DataFrame(all_combined_rows)
print(f"Combined DataFrame created: {combined_df.shape[0]} rows, {combined_df.shape[1]} columns")
```

```python
# Process flagged rows for RVer and Trucker availability
print("Processing flagged rows for RVer/Trucker availability...")

unique_source_rows = combined_df['df1_source_row'].unique()
updated_count = 0

for source_row in unique_source_rows:
    matching_rows = combined_df[combined_df['df1_source_row'] == source_row]
    
    # Check if there are exactly 2 matches
    if len(matching_rows) == 2:
        row_indices = matching_rows.index.tolist()
        stop_types = matching_rows['Scraped_stop_type'].tolist()
        
        # Check if one is RVer and the other is Trucker
        if (pd.notna(stop_types[0]) and pd.notna(stop_types[1]) and 
            set(stop_types) == {'RVer', 'Trucker'}):
            
            # Update both rows to unflag them
            for idx in row_indices:
                combined_df.loc[idx, 'Flagged'] = False
                combined_df.loc[idx, 'Flag_Reason'] = "Available in RVer and Trucker"
            updated_count += 2

print(f"Updated {updated_count} rows with RVer/Trucker availability")
print(f"Final flagged status: {combined_df['Flagged'].value_counts().to_dict()}")
```
