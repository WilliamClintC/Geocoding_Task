```python
# Filter for Address_Type = "Exit" and show OCR_address_standardized_OFF_parenthesis entries
exit_entries = df[df['Address_Type'] == 'Exit']
print(f"Found {len(exit_entries)} entries with Address_Type='Exit'")
print("\nEntries in OCR_address_standardized_OFF_parenthesis:")
print(exit_entries['OCR_address_standardized_OFF_parenthesis'].tolist())
```

```python
# Check which entries contain ")" characters
exit_entries_with_parenthesis = exit_entries[exit_entries['OCR_address_standardized_OFF_parenthesis'].str.contains('\)', na=False)]
print(f"Found {len(exit_entries_with_parenthesis)} entries containing ')' character:")
print("\nEntries with ')' character:")
for idx, entry in exit_entries_with_parenthesis['OCR_address_standardized_OFF_parenthesis'].items():
    print(f"Row {idx}: {entry}")
```

```python
# Remove ')' characters from OCR_address_standardized_OFF_parenthesis for Address_Type='Exit'
print("Before cleaning:")
for idx, entry in exit_entries_with_parenthesis['OCR_address_standardized_OFF_parenthesis'].items():
    print(f"Row {idx}: {entry}")

# Create a copy of the dataframe to modify
df_cleaned = df.copy()

# Remove ')' characters from entries where Address_Type='Exit'
mask = (df_cleaned['Address_Type'] == 'Exit')
df_cleaned.loc[mask, 'OCR_address_standardized_OFF_parenthesis'] = df_cleaned.loc[mask, 'OCR_address_standardized_OFF_parenthesis'].str.replace(')', '', regex=False)

print("\nAfter cleaning:")
# Show the cleaned entries
exit_entries_cleaned = df_cleaned[df_cleaned['Address_Type'] == 'Exit']
for idx in exit_entries_with_parenthesis.index:
    cleaned_entry = df_cleaned.loc[idx, 'OCR_address_standardized_OFF_parenthesis']
    print(f"Row {idx}: {cleaned_entry}")
```

```python
df= df_cleaned
df[['OCR_address_standardized_OFF_parenthesis', 'OCR_address_standardized_ON_parenthesis', 'Address_Type','OCR_zip_code','OCR_city','OCR_label']]
```
