```python
# Classification of Flag_Reason entries into three categories

# Define the three categories
successful_matches = [
    "Single Name/ZIP/State/Road Match Found (empty address type)",
    "Single Name/ZIP/State/Exit/Highway Match Found",
    "Available in RVer and Trucker",
]

multiple_matches = [
    "Multiple matches found",
    "Multiple Name/ZIP/State/Road Matches Found (empty address type) - Match 1 of 2",
    "Multiple Name/ZIP/State/Road Matches Found (empty address type) - Match 1 of 4",
    "Multiple Name/ZIP/State/Road Matches Found (empty address type) - Match 1 of 6",
    "Multiple Name/ZIP/State/Road Matches Found (empty address type) - Match 2 of 2",
    "Multiple Name/ZIP/State/Road Matches Found (empty address type) - Match 2 of 4",
    "Multiple Name/ZIP/State/Road Matches Found (empty address type) - Match 2 of 6",
    "Multiple Name/ZIP/State/Road Matches Found (empty address type) - Match 3 of 4",
    "Multiple Name/ZIP/State/Road Matches Found (empty address type) - Match 3 of 6",
    "Multiple Name/ZIP/State/Road Matches Found (empty address type) - Match 4 of 4",
    "Multiple Name/ZIP/State/Road Matches Found (empty address type) - Match 4 of 6",
    "Multiple Name/ZIP/State/Road Matches Found (empty address type) - Match 5 of 6",
    "Multiple Name/ZIP/State/Road Matches Found (empty address type) - Match 6 of 6",
    "Multiple Name/ZIP/State/Exit/Highway Matches Found - Match 1 of 2",
    "Multiple Name/ZIP/State/Exit/Highway Matches Found - Match 1 of 3",
    "Multiple Name/ZIP/State/Exit/Highway Matches Found - Match 1 of 4",
    "Multiple Name/ZIP/State/Exit/Highway Matches Found - Match 1 of 6",
    "Multiple Name/ZIP/State/Exit/Highway Matches Found - Match 2 of 2",
    "Multiple Name/ZIP/State/Exit/Highway Matches Found - Match 2 of 3",
    "Multiple Name/ZIP/State/Exit/Highway Matches Found - Match 2 of 4",
    "Multiple Name/ZIP/State/Exit/Highway Matches Found - Match 2 of 6",
    "Multiple Name/ZIP/State/Exit/Highway Matches Found - Match 3 of 3",
    "Multiple Name/ZIP/State/Exit/Highway Matches Found - Match 3 of 4",
    "Multiple Name/ZIP/State/Exit/Highway Matches Found - Match 3 of 6",
    "Multiple Name/ZIP/State/Exit/Highway Matches Found - Match 4 of 4",
    "Multiple Name/ZIP/State/Exit/Highway Matches Found - Match 4 of 6",
    "Multiple Name/ZIP/State/Exit/Highway Matches Found - Match 5 of 6",
    "Multiple Name/ZIP/State/Exit/Highway Matches Found - Match 6 of 6"
]

failed_matches = [
    "No Name/ZIP/State/Road matches found for empty address type",
    ""No Phone Match Found",
    "No matches found",
    "No Phone/Name/ZIP/State/Exit/Highway matches found",
    "no matching ZIPCODE",
    "matching ZIPCODE, matching State, no matching Exit",
    "matching ZIPCODE, matching State, matching Exit, no matching Highway/Exit/Street Address",
    "matching ZIPCODE, no matching State",
    "unclear OCR_address_standardized_OFF_parenthesis",
    "nan"  # missing data
]

# Get unique Flag_Reason values from the dataframe
unique_flag_reasons = df['Flag_Reason'].dropna().unique()

# Classify each unique flag reason
categorized_flags = {
    'Successful Matches': [],
    'Multiple Matches': [],
    'Failed Matches': [],
    'Unclassified': []
}

for flag in unique_flag_reasons:
    flag_str = str(flag).strip()
    
    if flag_str in successful_matches:
        categorized_flags['Successful Matches'].append(flag_str)
    elif flag_str in multiple_matches:
        categorized_flags['Multiple Matches'].append(flag_str)
    elif flag_str in failed_matches or flag_str == 'nan':
        categorized_flags['Failed Matches'].append(flag_str)
    else:
        categorized_flags['Unclassified'].append(flag_str)

# Display the results
print("=== FLAG REASON CLASSIFICATION ===\n")

for category, flags in categorized_flags.items():
    print(f"{category.upper()} ({len(flags)} entries):")
    print("-" * 50)
    for flag in sorted(flags):
        print(f"  • {flag}")
    print()

# Also handle NaN values separately
nan_count = df['Flag_Reason'].isna().sum()
if nan_count > 0:
    print(f"Missing/NaN Flag_Reason entries: {nan_count}")
    print()

# Summary statistics
total_entries = len(df)
print("=== SUMMARY STATISTICS ===")
print(f"Total records: {total_entries:,}")
print(f"Successful Matches: {len(df[df['Flag_Reason'].isin(successful_matches)]):,} ({len(df[df['Flag_Reason'].isin(successful_matches)])/total_entries*100:.1f}%)")
print(f"Multiple Matches: {len(df[df['Flag_Reason'].isin(multiple_matches)]):,} ({len(df[df['Flag_Reason'].isin(multiple_matches)])/total_entries*100:.1f}%)")
print(f"Failed Matches: {len(df[df['Flag_Reason'].isin(failed_matches + ['nan']) | df['Flag_Reason'].isna()]):,} ({len(df[df['Flag_Reason'].isin(failed_matches + ['nan']) | df['Flag_Reason'].isna()])/total_entries*100:.1f}%)")

# Check for any unclassified entries
unclassified_mask = ~(df['Flag_Reason'].isin(successful_matches + multiple_matches + failed_matches) | df['Flag_Reason'].isna())
unclassified_count = unclassified_mask.sum()
if unclassified_count > 0:
    print(f"Unclassified entries: {unclassified_count:,} ({unclassified_count/total_entries*100:.1f}%)")
    print("Unclassified Flag_Reason values:")
    for flag in df[unclassified_mask]['Flag_Reason'].unique():
        print(f"  • {flag}")
```

```python
# Concise Summary of Classification
print("=== CONCISE CLASSIFICATION SUMMARY ===\n")

# Count entries in each category
successful_count = len(df[df['Flag_Reason'].isin(successful_matches)])
multiple_count = len(df[df['Flag_Reason'].isin(multiple_matches)])
failed_count = len(df[df['Flag_Reason'].isin(failed_matches) | df['Flag_Reason'].isna()])
unclassified_count = len(df) - successful_count - multiple_count - failed_count

print(f"📊 DISTRIBUTION:")
print(f"✅ Successful Matches: {successful_count:,} records ({successful_count/len(df)*100:.1f}%)")
print(f"🔄 Multiple Matches: {multiple_count:,} records ({multiple_count/len(df)*100:.1f}%)")
print(f"❌ Failed Matches: {failed_count:,} records ({failed_count/len(df)*100:.1f}%)")

if unclassified_count > 0:
    print(f"❓ Unclassified: {unclassified_count:,} records ({unclassified_count/len(df)*100:.1f}%)")
    print(f"\n🔍 Unclassified Flag_Reason values:")
    unclassified_flags = df[~(df['Flag_Reason'].isin(successful_matches + multiple_matches + failed_matches) | df['Flag_Reason'].isna())]['Flag_Reason'].unique()
    for i, flag in enumerate(unclassified_flags[:10], 1):  # Show first 10
        print(f"   {i}. {flag}")
    if len(unclassified_flags) > 10:
        print(f"   ... and {len(unclassified_flags) - 10} more")

print(f"\n📈 Total Records: {len(df):,}")

# Create a new column with the categories
def classify_flag_reason(flag):
    if pd.isna(flag):
        return 'Failed Matches'
    flag_str = str(flag).strip()
    if flag_str in successful_matches:
        return 'Successful Matches'
    elif flag_str in multiple_matches:
        return 'Multiple Matches'
    elif flag_str in failed_matches or flag_str == 'nan':
        return 'Failed Matches'
    else:
        return 'Unclassified'

# Add the classification column to the dataframe
df['Flag_Category'] = df['Flag_Reason'].apply(classify_flag_reason)

print(f"\n✨ Added 'Flag_Category' column to the dataframe")
print(f"📋 Value counts for Flag_Category:")
print(df['Flag_Category'].value_counts())
```
