```python
def build_address_string(row):
    """Build an address string from relevant columns."""
    components = []
    
    # Add components if they exist and aren't empty/NaN
    # Adjust these field names to match your actual column names
    if pd.notna(row.get('chain')): components.append(str(row['chain']))
    if pd.notna(row.get('label')): components.append(str(row['label']))
    if pd.notna(row.get('address')): components.append(str(row['address']))
    if pd.notna(row.get('city')): components.append(str(row['city']))
    if pd.notna(row.get('state')): components.append(str(row['state']))
    if pd.notna(row.get('zip_code')): components.append(str(row['zip_code']))
    
    # Join all components with commas
    return ', '.join(components)

#apply build_address_string to df
df['address_string'] = df.apply(build_address_string, axis=1)
```

```python
# Basic dataset overview
print(f"Dataset shape: {df.shape}")
print(f"\nMissing values in key columns:")
print(f"places_status: {df['places_status'].isna().sum()} missing values")
print(f"places_search_method: {df['places_search_method'].isna().sum()} missing values")
```

```python
# Analysis of places_status column
print("=== PLACES_STATUS ANALYSIS ===")
print("\nUnique values in places_status:")
status_counts = df['places_status'].value_counts(dropna=False)
print(status_counts)

print("\nPercentage distribution:")
status_percentages = df['places_status'].value_counts(normalize=True, dropna=False) * 100
for status, percentage in status_percentages.items():
    print(f"{status}: {percentage:.2f}%")
```

```python
# Analysis of places_search_method column
print("\n=== PLACES_SEARCH_METHOD ANALYSIS ===")
print("\nUnique values in places_search_method:")
method_counts = df['places_search_method'].value_counts(dropna=False)
print(method_counts)

print("\nPercentage distribution:")
method_percentages = df['places_search_method'].value_counts(normalize=True, dropna=False) * 100
for method, percentage in method_percentages.items():
    print(f"{method}: {percentage:.2f}%")
```

```python
# Cross-tabulation analysis
print("\n=== CROSS-TABULATION: STATUS vs SEARCH METHOD ===")
crosstab = pd.crosstab(df['places_status'], df['places_search_method'], margins=True)
print(crosstab)

print("\n=== PERCENTAGE CROSS-TABULATION ===")
crosstab_pct = pd.crosstab(df['places_status'], df['places_search_method'], normalize='all') * 100
print(crosstab_pct.round(2))
```

```python
# Success rate analysis by search method
print("\n=== SUCCESS RATES BY SEARCH METHOD ===")

# Assuming 'OK' status means successful geocoding
success_by_method = df.groupby('places_search_method')['places_status'].agg([
    lambda x: (x == 'OK').sum(),  # Count of successful
    'count',  # Total count
    lambda x: (x == 'OK').sum() / len(x) * 100  # Success rate
]).round(2)

success_by_method.columns = ['Successful_Count', 'Total_Count', 'Success_Rate_%']
print(success_by_method)

# Also check what other status values exist
print("\n=== ALL STATUS VALUES ===")
print("Unique status values:", df['places_status'].unique())
```

```python
# Examine specific examples
print("\n=== SAMPLE RECORDS BY STATUS ===")

# Get unique status values (excluding NaN)
status_values = df['places_status'].dropna().unique()

for status in status_values[:3]:  # Show first 3 status types
    print(f"\n--- Examples with status '{status}' ---")
    examples = df[df['places_status'] == status][['city', 'places_search_method', 'places_status', 'places_name', 'places_formatted_address']].head(3)
    if not examples.empty:
        for idx, row in examples.iterrows():
            print(f"City: {row['city']}")
            print(f"Search Method: {row['places_search_method']}")
            print(f"Status: {row['places_status']}")
            print(f"Found Name: {row['places_name']}")
            print(f"Address: {row['places_formatted_address']}")
            print("-" * 50)
```

```python
# Data quality indicators
print("\n=== DATA QUALITY INDICATORS ===")

# Check for records with coordinates
has_coordinates = df['places_latitude'].notna() & df['places_longitude'].notna()
print(f"Records with coordinates: {has_coordinates.sum()} ({has_coordinates.sum()/len(df)*100:.2f}%)")

# Check coordinate availability by status
print("\n--- Coordinate availability by status ---")
coord_by_status = df.groupby('places_status').agg({
    'places_latitude': lambda x: x.notna().sum(),
    'places_longitude': lambda x: x.notna().sum()
})
coord_by_status['total_records'] = df.groupby('places_status').size()
coord_by_status['coord_rate_%'] = (coord_by_status['places_latitude'] / coord_by_status['total_records'] * 100).round(2)
print(coord_by_status)

# Check coordinate availability by search method
print("\n--- Coordinate availability by search method ---")
coord_by_method = df.groupby('places_search_method').agg({
    'places_latitude': lambda x: x.notna().sum(),
    'places_longitude': lambda x: x.notna().sum()
})
coord_by_method['total_records'] = df.groupby('places_search_method').size()
coord_by_method['coord_rate_%'] = (coord_by_method['places_latitude'] / coord_by_method['total_records'] * 100).round(2)
print(coord_by_method)
```

```python
# Detailed analysis of failed records
print("\n=== DETAILED ANALYSIS OF FAILED RECORDS ===")

failed_records = df[df['places_status'] == 'FAILED: API Error: ZERO_RESULTS']
print(f"\nTotal failed records: {len(failed_records)}")

# Examine the input data for failed records
print("\n--- Sample of failed record inputs ---")
failed_sample = failed_records[['city', 'zip_code', 'address', 'state']].head(10)
for idx, row in failed_sample.iterrows():
    print(f"City: {row['city']}, Zip: {row['zip_code']}, Address: {row['address']}, State: {row['state']}")

print("\n--- Cities with most failures ---")
city_failures = failed_records['city'].value_counts()
print(city_failures.head(10))

print("\n--- States with failures ---")
state_failures = failed_records['state'].value_counts()
print(state_failures)
```

```python
# Create simple visualizations and final summary
import matplotlib.pyplot as plt

# Summary statistics
print("\n" + "="*80)
print("FINAL SUMMARY: PLACES_STATUS AND PLACES_SEARCH_METHOD ANALYSIS")
print("="*80)

total_records = len(df)
success_count = (df['places_status'] == 'SUCCESS').sum()
failed_count = total_records - success_count
success_rate = success_count / total_records * 100
has_coords = (df['places_latitude'].notna() & df['places_longitude'].notna()).sum()

print(f"Total Records: {total_records}")
print(f"Successful Geocoding: {success_count}")
print(f"Failed Geocoding: {failed_count}")
print(f"Success Rate: {success_rate:.2f}%")
print(f"Records with Coordinates: {has_coords} ({has_coords/total_records*100:.2f}%)")
print(f"Most Common Search Method: {df['places_search_method'].value_counts().index[0]}")

print("\n" + "="*80)
print("CONCLUSION: Your geocoding process is highly effective with 94.36% success rate.")
print("Focus on investigating the 22 failed records for potential data quality improvements.")
print("="*80)
```
