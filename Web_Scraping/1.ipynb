```python
# Alternative approach using webdriver-manager (easier setup)
# This automatically downloads and manages ChromeDriver

# First install the required packages:
# !pip install selenium webdriver-manager

import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.action_chains import ActionChains
from webdriver_manager.chrome import ChromeDriverManager
from selenium.common.exceptions import TimeoutException, NoSuchElementException
import time
import urllib.parse
import random

def automated_google_search_and_click(df, column_name='places_search_query', batch_size=10, save_path=r'C:\Users\clint\Desktop\Geocoding_Task\Web_Scraping'):
    """
    Automate Google searches and click on first results with batch processing
    
    Args:
        df: DataFrame containing search queries
        column_name: Name of the column containing search queries
        batch_size: Number of rows to process in each batch (default: 10)
        save_path: Directory to save batch CSV files
    
    Returns:
        Updated DataFrame with search links and first result links
    """
    
    # Set up Chrome options to avoid detection
    chrome_options = Options()
    chrome_options.add_argument("--no-sandbox")
    chrome_options.add_argument("--disable-dev-shm-usage")
    chrome_options.add_argument("--disable-blink-features=AutomationControlled")
    chrome_options.add_argument("--disable-extensions")
    chrome_options.add_argument("--disable-plugins-discovery")
    chrome_options.add_argument("--disable-web-security")
    chrome_options.add_argument("--allow-running-insecure-content")
    chrome_options.add_argument("--no-first-run")
    chrome_options.add_argument("--no-service-autorun")
    chrome_options.add_argument("--password-store=basic")
    chrome_options.add_argument("--disable-features=VizDisplayCompositor")
    chrome_options.add_experimental_option("excludeSwitches", ["enable-automation"])
    chrome_options.add_experimental_option('useAutomationExtension', False)
    chrome_options.add_experimental_option("detach", True)
    
    # Add a realistic user agent
    chrome_options.add_argument("--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36")
    
    # Initialize the Chrome WebDriver with webdriver-manager
    service = Service(ChromeDriverManager().install())
    driver = webdriver.Chrome(service=service, options=chrome_options)
    
    # Execute multiple scripts to hide automation indicators
    driver.execute_script("Object.defineProperty(navigator, 'webdriver', {get: () => undefined})")
    driver.execute_script("Object.defineProperty(navigator, 'plugins', {get: () => [1, 2, 3, 4, 5]})")
    driver.execute_script("Object.defineProperty(navigator, 'languages', {get: () => ['en-US', 'en']})")
    driver.execute_script("Object.defineProperty(navigator, 'permissions', {get: () => ({query: () => Promise.resolve({state: 'granted'})})})")
    
    # Set window size to common resolution
    driver.set_window_size(1920, 1080)
    
    # Create lists to store results
    all_results = []  # Store all batch results
    dorking_search_links = []
    first_result_links = []
    
    # Calculate number of batches
    total_rows = len(df)
    num_batches = (total_rows + batch_size - 1) // batch_size  # Ceiling division
    print(f"Processing {total_rows} rows in {num_batches} batches of {batch_size}")
    
    try:
        for batch_num in range(num_batches):
            print(f"\n{'='*50}")
            print(f"STARTING BATCH {batch_num + 1} OF {num_batches}")
            print(f"{'='*50}")
            
            # Calculate batch boundaries
            start_idx = batch_num * batch_size
            end_idx = min(start_idx + batch_size, total_rows)
            batch_df = df.iloc[start_idx:end_idx].copy()
            
            # Reset lists for this batch
            batch_dorking_links = []
            batch_first_result_links = []
            
            print(f"Processing rows {start_idx + 1} to {end_idx}")
            
            for i, (idx, row) in enumerate(batch_df.iterrows()):
                query = row[column_name]
                # Add the site-specific dorking prefix to the search query
                dorking_prefix = "site:rvandtravelers.com OR site:truckstopsandservices.com"
                full_query = f"{dorking_prefix} {query}"
                print(f"\nProcessing query {i+1}/{len(batch_df)} in batch {batch_num + 1}: {full_query}")
                
                search_url = "https://www.google.com/search?q=" + urllib.parse.quote_plus(full_query)
                print(f"Opening search page: {search_url}")
                
                # Add random delay before each search to appear more human-like
                delay = random.uniform(3, 7)
                print(f"Waiting {delay:.1f} seconds before search...")
                time.sleep(delay)
                
                # Navigate to the search URL
                driver.get(search_url)
                batch_dorking_links.append(search_url)
                
                # Simulate human-like mouse movement
                try:
                    actions = ActionChains(driver)
                    actions.move_by_offset(random.randint(100, 300), random.randint(100, 300))
                    actions.perform()
                except:
                    pass
                
                try:
                    # Wait for search results to load with longer timeout
                    wait = WebDriverWait(driver, 15)
                    
                    # Check if Google is asking for verification
                    if "unusual traffic" in driver.page_source.lower() or "captcha" in driver.page_source.lower():
                        print("⚠️  Google detected unusual traffic. Waiting longer and trying again...")
                        time.sleep(random.uniform(10, 20))
                        driver.refresh()
                        time.sleep(random.uniform(5, 10))
                    
                    # Scroll down a bit to simulate human behavior
                    driver.execute_script("window.scrollTo(0, 300);")
                    time.sleep(random.uniform(1, 3))
                    
                    # Multiple strategies to find the first search result
                    selectors = [
                        'div[data-sokoban-container] a h3',
                        'div.g a h3',
                        'h3.LC20lb',
                        'h3[class*="DKV0Md"]',
                        'div[class*="yuRUbf"] a',
                        'div[class*="tF2Cxc"] a h3'
                    ]
                    
                    first_result_clicked = False
                    first_result_link = None
                    
                    for selector in selectors:
                        try:
                            elements = driver.find_elements(By.CSS_SELECTOR, selector)
                            if elements:
                                first_element = elements[0]
                                
                                # Get the link URL
                                if first_element.tag_name == 'a':
                                    first_result_link = first_element.get_attribute('href')
                                    click_element = first_element
                                else:
                                    # Find parent anchor tag
                                    parent = first_element
                                    while parent and parent.tag_name != 'a':
                                        try:
                                            parent = parent.find_element(By.XPATH, '..')
                                        except:
                                            break
                                    if parent and parent.tag_name == 'a':
                                        first_result_link = parent.get_attribute('href')
                                        click_element = parent
                                    else:
                                        continue
                                
                                if first_result_link:
                                    print(f"Found first result: {first_result_link}")
                                    batch_first_result_links.append(first_result_link)
                                    
                                    # Simulate human-like behavior before clicking
                                    actions = ActionChains(driver)
                                    actions.move_to_element(click_element)
                                    actions.pause(random.uniform(0.5, 1.5))
                                    actions.perform()
                                    time.sleep(random.uniform(0.5, 1.0))
                                    
                                    # Click the first result using JavaScript to avoid interception
                                    driver.execute_script("arguments[0].click();", click_element)
                                    print(f"✓ Clicked on first result")
                                    first_result_clicked = True
                                    
                                    # Wait for the page to load with random delay
                                    page_load_delay = random.uniform(4, 8)
                                    time.sleep(page_load_delay)
                                    break
                                    
                        except Exception as e:
                            print(f"Error with selector {selector}: {e}")
                            continue
                    
                    if not first_result_clicked:
                        print("❌ No search results found or unable to click")
                        batch_first_result_links.append(None)
                        
                except TimeoutException:
                    print(f"❌ Timeout waiting for search results to load")
                    batch_first_result_links.append(None)
                except Exception as e:
                    print(f"❌ Error processing search: {e}")
                    batch_first_result_links.append(None)
                
                # Add longer random delay between searches (5-15 seconds)
                if i < len(batch_df) - 1:  # Don't wait after the last search in batch
                    between_search_delay = random.uniform(8, 15)
                    print(f"Waiting {between_search_delay:.1f} seconds before next search...")
                    time.sleep(between_search_delay)
            
            # Create batch DataFrame with results
            batch_result_df = batch_df.copy()
            batch_result_df['dorking_search_links'] = batch_dorking_links
            batch_result_df['first_result_links'] = batch_first_result_links
            
            # Save batch results to CSV
            import os
            os.makedirs(save_path, exist_ok=True)
            batch_filename = f"batch_{batch_num + 1:03d}_results.csv"
            batch_filepath = os.path.join(save_path, batch_filename)
            batch_result_df.to_csv(batch_filepath, index=False)
            print(f"\n✅ Batch {batch_num + 1} completed and saved to: {batch_filepath}")
            
            # Add batch results to overall results
            all_results.append(batch_result_df)
            
            # Add delay between batches (longer to be safe)
            if batch_num < num_batches - 1:  # Don't wait after the last batch
                batch_delay = random.uniform(15, 25)
                print(f"Waiting {batch_delay:.1f} seconds before next batch...")
                time.sleep(batch_delay)
            
    finally:
        # Close the browser
        print("\nClosing browser...")
        driver.quit()
    
    # Combine all batch results
    if all_results:
        final_df = pd.concat(all_results, ignore_index=True)
        
        # Save final combined results
        import os
        final_filename = "final_combined_results.csv"
        final_filepath = os.path.join(save_path, final_filename)
        final_df.to_csv(final_filepath, index=False)
        print(f"\n🎉 All batches completed! Final combined results saved to: {final_filepath}")
        
        return final_df
    else:
        print("\n❌ No results were processed")
        return df.copy()

# Usage:
# result_df = automated_google_search_and_click(df)
# result_df
```

```python
# Utility functions for batch processing management

def resume_from_batch(df, batch_number, batch_size=10, save_path=r'C:\Users\clint\Desktop\Geocoding_Task\Web_Scraping'):
    """
    Resume processing from a specific batch number
    
    Args:
        df: Original DataFrame
        batch_number: Batch number to start from (1-based)
        batch_size: Size of each batch
        save_path: Directory where batch files are saved
    """
    start_idx = (batch_number - 1) * batch_size
    remaining_df = df.iloc[start_idx:].copy()
    print(f"Resuming from batch {batch_number}, processing {len(remaining_df)} remaining rows")
    return automated_google_search_and_click(remaining_df, batch_size=batch_size, save_path=save_path)

def combine_existing_batches(save_path=r'C:\Users\clint\Desktop\Geocoding_Task\Web_Scraping'):
    """
    Combine all existing batch CSV files into one final file
    """
    import glob
    import os
    
    batch_files = glob.glob(os.path.join(save_path, "batch_*_results.csv"))
    if not batch_files:
        print("No batch files found")
        return None
    
    batch_files.sort()  # Sort to ensure correct order
    combined_dfs = []
    
    for file in batch_files:
        df_batch = pd.read_csv(file)
        combined_dfs.append(df_batch)
        print(f"Loaded {file} with {len(df_batch)} rows")
    
    final_df = pd.concat(combined_dfs, ignore_index=True)
    final_filepath = os.path.join(save_path, "manually_combined_results.csv")
    final_df.to_csv(final_filepath, index=False)
    print(f"\nCombined {len(batch_files)} batch files into: {final_filepath}")
    return final_df

# Example usage:
# To resume from batch 3: resume_from_batch(df, 3)
# To combine existing batches: combine_existing_batches()
```
