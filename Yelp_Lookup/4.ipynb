```python
# Normalize phone numbers in df and df2 for comparison
import re

# Function to normalize phone numbers to a standard format
def normalize_phone(phone_num):
    if pd.isna(phone_num):
        return None
    
    # Remove all non-digit characters
    digits_only = re.sub(r'\D', '', str(phone_num))
    
    # If it's a 10-digit number, add '1' prefix for US numbers
    if len(digits_only) == 10:
        return '1' + digits_only
    
    # If it already has 11 digits starting with 1, return as is
    elif len(digits_only) == 11 and digits_only.startswith('1'):
        return digits_only
    
    # Return whatever we have if it doesn't fit the expected patterns
    return digits_only

# Create a new column for the matching row ids
df['Phone_Yelp_matches_row_ids'] = None

# Normalize the phone numbers in both dataframes
df['normalized_phone'] = df['phone'].apply(normalize_phone)
df2['normalized_phone'] = df2['Original_Phone'].apply(normalize_phone)

# Create a dictionary mapping normalized phone numbers to their row indices in df2
# But only include rows where df2['Name'] is not missing
phone_to_indices = {}
for idx, row in df2.iterrows():
    phone = row['normalized_phone']
    # Only include rows where Name is not missing
    if phone is not None and not pd.isna(row['Name']):
        if phone not in phone_to_indices:
            phone_to_indices[phone] = []
        phone_to_indices[phone].append(idx)

# For each row in df, find matching rows in df2
for idx, row in df.iterrows():
    if row['normalized_phone'] is not None:
        # Check if the normalized phone exists in our mapping
        if row['normalized_phone'] in phone_to_indices:
            df.at[idx, 'Phone_Yelp_matches_row_ids'] = phone_to_indices[row['normalized_phone']]

# Clean up by removing the temporary normalized_phone column
df.drop('normalized_phone', axis=1, inplace=True)
df2.drop('normalized_phone', axis=1, inplace=True)

# Display the number of matches found
match_count = df['Phone_Yelp_matches_row_ids'].notna().sum()
print(f"Number of rows in df with matching phone numbers in df2 (where df2['Name'] is not missing): {match_count}")

# Preview the matched results
print("\nPreview of rows with matches:")
matched_df = df[df['Phone_Yelp_matches_row_ids'].notna()]
display(matched_df.head())
```

```python
# Examine some of the matches to verify they are correct
if match_count > 0:
    print("Sample of matched phone numbers (only matches where df2['Name'] is not missing):")
    sample_size = min(5, match_count)  # Show up to 5 matches
    
    for i, (idx, row) in enumerate(matched_df.head(sample_size).iterrows()):
        print(f"\nMatch {i+1}:")
        print(f"df row {idx}:")
        # Display df['Name'] if it exists
        if 'Name' in df.columns:
            print(f"  - Name: {row['Name']}")
        print(f"  - Phone: {row['phone']}")
        
        # Get the matched row ids from df2
        matched_ids = row['Phone_Yelp_matches_row_ids']
        if isinstance(matched_ids, list):
            print(f"  - Found {len(matched_ids)} matches in Yelp data:")
            for match_id in matched_ids:
                matched_row = df2.iloc[match_id]
                print(f"    df2 row {match_id}:")
                print(f"      - Name: {matched_row['Name']}")  # We know Name exists and is not NA
                print(f"      - Phone: {matched_row['Original_Phone']}")
        else:
            # In case there's only one match and it's not stored as a list
            matched_row = df2.iloc[matched_ids]
            print(f"  - Found 1 match in Yelp data:")
            print(f"    df2 row {matched_ids}:")
            print(f"      - Name: {matched_row['Name']}")  # We know Name exists and is not NA
            print(f"      - Phone: {matched_row['Original_Phone']}")
else:
    print("No matches found between the dataframes where df2['Name'] is not missing.")
```
