```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import ast

```

```python


df = pd.read_csv(r'C:\Users\clint\Desktop\Geocoding_Task\Yelp_Lookup\7_5.csv')
df
```

```python
# Target sequence to follow
target_sequence = [2006, 2007, 2008, 2014, 2015, 2016]

# Initialize
n = 1
group_ids = [None] * len(df)
i = 0  # Current index in df

while i < len(df):
    # Process one row at a time
    current_year = df.at[i, 'year']
    
    # Check if current year is in target sequence (can start with any year)
    if current_year in target_sequence:
        # Find where this year appears in the sequence
        seq_idx = target_sequence.index(current_year)
        
        # Start a new group
        group_ids[i] = n
        seq_idx += 1
        i += 1
        
        # Continue building this group while we find matching years
        while i < len(df) and seq_idx < len(target_sequence):
            current_year = df.at[i, 'year']
            
            # If we find the next expected year, add to current group
            if current_year == target_sequence[seq_idx]:
                group_ids[i] = n
                seq_idx += 1
                i += 1
            else:
                # If we don't find the expected year, break out and start new group
                break
        
        # Move to next group after completing or breaking current group
        n += 1
    else:
        # If current row doesn't match any year in sequence, skip it (no group assignment)
        i += 1

df['place_identifier(year)'] = group_ids
```

```python
# Check which place_identifier(year) groups don't have exactly 6 entries
place_counts = df['place_identifier(year)'].value_counts().sort_index()
print("Count of entries for each place_identifier(year):")
print(place_counts)
print("\n" + "="*50)

# Find place identifiers that don't have exactly 6 entries
not_six_entries = place_counts[place_counts != 6]
print(f"\nPlace identifiers that DON'T have exactly 6 entries:")
if len(not_six_entries) > 0:
    for place_id, count in not_six_entries.items():
        print(f"Place identifier {place_id}: {count} entries")
        # Show the actual rows for this place identifier
        print(f"Rows for place identifier {place_id}:")
        place_rows = df[df['place_identifier(year)'] == place_id][['year', 'place_identifier(year)']]
        print(place_rows)
        print("-" * 30)
else:
    print("All place identifiers have exactly 6 entries!")
```

```python
# Summary analysis
total_places = df['place_identifier(year)'].nunique()
places_with_6_entries = (place_counts == 6).sum()
places_without_6_entries = (place_counts != 6).sum()

print(f"SUMMARY:")
print(f"Total unique place identifiers: {total_places}")
print(f"Place identifiers with exactly 6 entries: {places_with_6_entries}")
print(f"Place identifiers WITHOUT exactly 6 entries: {places_without_6_entries}")

# Show distribution of entry counts
print(f"\nDistribution of entry counts:")
count_distribution = place_counts.value_counts().sort_index()
for count, freq in count_distribution.items():
    print(f"{freq} place identifiers have {count} entries each")

# Show total rows
print(f"\nTotal rows in dataset: {len(df)}")
print(f"Rows with missing place_identifier: {df['place_identifier(year)'].isnull().sum()}")
```

```python
# Create a DataFrame showing all places that don't have exactly 6 entries
problematic_places = []

for place_id, count in not_six_entries.items():
    # Get all rows for this place identifier
    place_data = df[df['place_identifier(year)'] == place_id].copy()
    
    # Add summary info for each place
    for _, row in place_data.iterrows():
        problematic_places.append({
            'place_identifier': place_id,
            'entry_count': count,
            'year': row['year'],
            'missing_years': 6 - count
        })

# Convert to DataFrame
problematic_df = pd.DataFrame(problematic_places)

print("DataFrame of all place identifiers that DON'T have exactly 6 entries:")
print("="*70)
problematic_df
```

```python
# Group and display by entry count for better organization
print("DETAILED BREAKDOWN BY ENTRY COUNT:")
print("="*50)

for entry_count in sorted(problematic_df['entry_count'].unique()):
    places_with_this_count = problematic_df[problematic_df['entry_count'] == entry_count]['place_identifier'].unique()
    print(f"\nPlaces with {entry_count} entries ({len(places_with_this_count)} places):")
    print("-" * 30)
    
    for place_id in places_with_this_count:
        place_subset = problematic_df[problematic_df['place_identifier'] == place_id]
        years = sorted(place_subset['year'].tolist())
        missing_count = place_subset['missing_years'].iloc[0]
        
        print(f"Place ID {place_id}: Years = {years} (Missing {missing_count} years)")
        
        # Show which years are missing from the target sequence
        target_set = set([2006, 2007, 2008, 2014, 2015, 2016])
        present_years = set(years)
        missing_years = sorted(list(target_set - present_years))
        if missing_years:
            print(f"   Missing years: {missing_years}")
        print()
```

```python
# Create a clean summary DataFrame
summary_data = []

for entry_count in sorted(problematic_df['entry_count'].unique()):
    places_with_this_count = problematic_df[problematic_df['entry_count'] == entry_count]['place_identifier'].unique()
    
    for place_id in places_with_this_count:
        place_subset = problematic_df[problematic_df['place_identifier'] == place_id]
        years = sorted(place_subset['year'].tolist())
        
        # Calculate missing years
        target_set = set([2006, 2007, 2008, 2014, 2015, 2016])
        present_years = set(years)
        missing_years = sorted(list(target_set - present_years))
        
        summary_data.append({
            'place_identifier': place_id,
            'entry_count': entry_count,
            'years_present': years,
            'years_missing': missing_years,
            'missing_count': len(missing_years)
        })

summary_df = pd.DataFrame(summary_data)

print("\nCLEAN SUMMARY DATAFRAME:")
print("="*40)
summary_df
```
