```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import ast

```

```python


df = pd.read_csv(r'C:\Users\clint\Desktop\Geocoding_Task\Yelp_Lookup\8.csv')
df
```

```python
# Define the enhanced combined condition
enhanced_combined_condition = (
    ((df['Scraped_phone_match_rate'] == True) | 
     (df['Scraped_zipcode_to_label_match_rate'] == '6/6 successful match') | 
     (df['Scraped_zipcode_to_label_match_rate'] == '7/6 successful match')) |
    (df['Yelp_phone_match_rate'] == True)
)

print("=== ENHANCED COMBINED CONDITION ANALYSIS ===")
print(f"Total rows in dataset: {len(df)}")

# Count rows that meet the enhanced criteria
enhanced_success_count = enhanced_combined_condition.sum()
print(f"Rows meeting enhanced criteria: {enhanced_success_count} ({enhanced_success_count/len(df)*100:.2f}%)")

# Count rows that DON'T meet the enhanced criteria (what we want to isolate)
enhanced_low_success_condition = ~enhanced_combined_condition
enhanced_low_success_count = enhanced_low_success_condition.sum()
print(f"Rows NOT meeting enhanced criteria: {enhanced_low_success_count} ({enhanced_low_success_count/len(df)*100:.2f}%)")

# Isolate the rows that don't meet the criteria
isolated_rows = df[enhanced_low_success_condition].copy()
print(f"\nIsolated {len(isolated_rows)} rows that don't meet the enhanced criteria")

# Display basic info about the isolated rows
print(f"\nShape of isolated dataset: {isolated_rows.shape}")
isolated_rows.head()
```
