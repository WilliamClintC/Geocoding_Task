```python
# Check unique values of OCR_Address_Type and their counts
print("Unique OCR_Address_Type values and counts:")
print(df['OCR_Address_Type'].value_counts())
print("\nRows where OCR_Address_Type == 'empty':")
empty_rows = df[df['OCR_Address_Type'] == 'empty']
print(f"Count: {len(empty_rows)}")
print("\nSample of empty rows:")
print(empty_rows[['OCR_Address_Type', 'OCR_zip_code', 'OCR_label', 'OCR_Main_Road', 'Flagged', 'Flag_Reason']].head())

################################################################################
# MODIFIED MATCHING LOGIC FOR OCR_Address_Type=="empty"
################################################################################

import pandas as pd
import numpy as np
import re

def clean_text_for_matching(text):
    """Clean text for case-insensitive matching, removing hyphens, apostrophes, spaces, periods"""
    if pd.isna(text):
        return ""
    return re.sub(r"['-.\s]", "", str(text).lower())

def filter_meaningful_words(text):
    """
    Extract meaningful words for name matching, excluding common generic terms.
    Returns a list of words that are meaningful for matching business names.
    """
    if pd.isna(text):
        return []
    
    # Common words to exclude from matching (case-insensitive)
    exclude_words = {
        # Vehicle/Transportation related
        'truck', 'trucks', 'trucker', 'truckers', 'trucking',
        'stop', 'stops', 'station', 'stations',
        'travel', 'travels', 'traveling', 'centre', 'center',
        'plaza', 'plazas', 'mart', 'store', 'stores',
        'gas', 'fuel', 'petrol', 'petroleum', 'diesel',
        'service', 'services', 'auto', 'automotive',
        'parking', 'fleet', 'commercial', 'professional',
        
        # Generic business terms
        'inc', 'llc', 'corp', 'corporation', 'company', 'co',
        'ltd', 'limited', 'enterprises', 'enterprise',
        'group', 'international', 'intl', 'usa', 'america',
        'franchise', 'chain', 'brand', 'corporate',
        
        # Location/Direction terms (but keep specific place names)
        'north', 'south', 'east', 'west', 'ne', 'nw', 'se', 'sw',
        'highway', 'hwy', 'road', 'rd', 'street', 'st', 'avenue', 'ave',
        'exit', 'interchange', 'junction',
        
        # Common adjectives/descriptors
        'big', 'small', 'large', 'giant', 'super', 'mega', 'express',
        'quick', 'fast', 'convenient', 'friendly', 'family',
        'country', 'state', 'national', 'speedy', 'instant',
        
        # Food/Restaurant terms
        'restaurant', 'cafe', 'diner', 'grill', 'kitchen', 'food', 'eats',
        
        # Accommodation terms
        'motel', 'hotel', 'inn', 'lodge', 'rest',
        
        # Time/Convenience terms
        '24', 'hour', 'hours',
        
        # Connectors and articles
        'and', 'or', 'the', 'of', 'in', 'at', 'on', 'for', 'with', 'by'
    }
    
    # Split text into words and clean each word
    words = str(text).lower().split()
    meaningful_words = []
    
    for word in words:
        # Clean the word (remove punctuation, etc.)
        cleaned_word = re.sub(r"[^\w]", "", word)
        
        # Keep words that are:
        # 1. At least 3 characters long
        # 2. Not in the exclude list
        # 3. Not purely numeric
        if (len(cleaned_word) >= 3 and 
            cleaned_word not in exclude_words and 
            not cleaned_word.isdigit()):
            meaningful_words.append(cleaned_word)
    
    return meaningful_words

def normalize_zip_code(zip_code):
    """Normalize ZIP code to 5 digits with leading zeros (e.g., '2345' -> '02345')"""
    if pd.isna(zip_code):
        return None
    try:
        # Remove any non-digit characters and convert to string
        zip_str = re.sub(r'\D', '', str(zip_code))
        if len(zip_str) >= 5:
            return zip_str[:5]  # Take first 5 digits
        elif len(zip_str) > 0:
            return zip_str.zfill(5)  # Pad with leading zeros to make 5 digits
        else:
            return None
    except:
        return None

def comprehensive_matching_logic_for_empty():
    """
    Modified implementation for OCR_Address_Type=="empty" rows.
    Since "empty" rows may not have exit numbers, we'll use a more flexible matching approach:
    1. Filter rows where OCR_Address_Type equals "empty" (no Flag_Reason restriction)
    2. Progressive matching: ZIP -> State -> Highway/Road -> Name (no exit number matching)
    3. Handle different match scenarios and create duplicate rows for multiple matches
    """
    
    # Step 1: Filter rows for "empty" address type only
    filtered_df = df[df['OCR_Address_Type'] == "empty"].copy()
    
    if len(filtered_df) == 0:
        print("No rows found matching the filter criteria (OCR_Address_Type=='empty')")
        return df
    
    # Step 2: Process all filtered rows
    selected_rows = filtered_df.copy()
    
    # Create working copy of dataframe
    result_df = df.copy()
    
    # Initialize "Manually Verified?" column with default value "No"
    if 'Manually Verified?' not in result_df.columns:
        result_df['Manually Verified?'] = "No"
    
    # Keep track of new rows to add
    new_rows_to_add = []
    
    # Prepare df2 with normalized ZIP codes for efficient matching
    df2_work = df2.copy()
    df2_work['ZIP_normalized'] = df2_work['Postal Code'].apply(normalize_zip_code)
    
    # Create mapping from df2 columns to Scraped_ columns
    def map_df2_to_scraped(df2_row):
        """Map a df2 row to scraped column values"""
        scraped_data = {}
        for df2_col in df2.columns:
            scraped_col = f"Scraped_{df2_col}"
            if scraped_col in result_df.columns:
                scraped_data[scraped_col] = df2_row[df2_col]
        return scraped_data
    
    print(f"Processing {len(selected_rows)} rows with OCR_Address_Type=='empty'...")
    
    # Process each selected row
    for row_num, (idx, row) in enumerate(selected_rows.iterrows(), 1):
        print(f"\nRow {row_num}/{len(selected_rows)} (Index {idx}):")
        print(f"  OCR Data: ZIP={row['OCR_zip_code']}, State={row['OCR_state']}")
        print(f"  Roads: Main='{row['OCR_Main_Road']}', Secondary='{row['OCR_Secondary_Road']}'")
        print(f"  Label: '{row['OCR_label']}'")
        
        # Step 3: Extract and normalize OCR_zip_code
        ocr_zip = normalize_zip_code(row['OCR_zip_code'])
        
        # Step 4: Check ZIP code match
        if not ocr_zip:
            result_df.loc[idx, 'Flagged'] = True
            result_df.loc[idx, 'Flag_Reason'] = "no matching ZIPCODE"
            print(f"  ❌ No valid ZIP code")
            continue
            
        zip_matches = df2_work[df2_work['ZIP_normalized'] == ocr_zip]
        if zip_matches.empty:
            result_df.loc[idx, 'Flagged'] = True
            result_df.loc[idx, 'Flag_Reason'] = "no matching ZIPCODE"
            print(f"  ❌ ZIP {ocr_zip} not found")
            continue
        
        print(f"  ✓ ZIP {ocr_zip}: {len(zip_matches)} matches")
        
        # Step 5: Check state match
        ocr_state = row['OCR_state']
        state_matches = zip_matches[zip_matches['State'] == ocr_state]
        if state_matches.empty:
            result_df.loc[idx, 'Flagged'] = True
            result_df.loc[idx, 'Flag_Reason'] = "matching ZIPCODE, no matching State"
            print(f"  ❌ State {ocr_state} not found")
            continue
        
        print(f"  ✓ State {ocr_state}: {len(state_matches)} matches")
        
        # Step 6: Comprehensive road matching
        # Check both OCR_Main_Road and OCR_Secondary_Road against all df2 road columns
        ocr_main_road = row['OCR_Main_Road']
        ocr_secondary_road = row['OCR_Secondary_Road']
        
        # Start with all state matches
        road_matches = state_matches.copy()
        
        # Collect all OCR road values that have data
        ocr_road_values = []
        if pd.notna(ocr_main_road) and str(ocr_main_road).strip():
            ocr_road_values.append(('Main Road', str(ocr_main_road).strip()))
        if pd.notna(ocr_secondary_road) and str(ocr_secondary_road).strip():
            ocr_road_values.append(('Secondary Road', str(ocr_secondary_road).strip()))
        
        # If we have any road information, try to filter by it
        if ocr_road_values:
            print(f"  🔍 Road matching using: {[f'{label}: {value}' for label, value in ocr_road_values]}")
            
            # Create comprehensive road matching condition
            # Check each OCR road value against all df2 road columns
            all_road_conditions = []
            
            for road_label, road_value in ocr_road_values:
                print(f"    🔍 Checking {road_label}: '{road_value}'")
                
                # Check against all df2 road-related columns
                road_condition = (
                    # Contains matching (partial)
                    (state_matches['Highway'].astype(str).str.contains(road_value, case=False, na=False, regex=False)) |
                    (state_matches['Exit'].astype(str).str.contains(road_value, case=False, na=False, regex=False)) |
                    (state_matches['Street Address'].astype(str).str.contains(road_value, case=False, na=False, regex=False)) |
                    (state_matches['Road Name'].astype(str).str.contains(road_value, case=False, na=False, regex=False)) |
                    # Exact matching
                    (state_matches['Highway'] == road_value) |
                    (state_matches['Exit'] == road_value) |
                    (state_matches['Street Address'] == road_value) |
                    (state_matches['Road Name'] == road_value)
                )
                
                all_road_conditions.append(road_condition)
                
                # Check how many matches this road value found
                road_matches_for_this_value = state_matches[road_condition]
                print(f"      📊 {road_label} '{road_value}' found {len(road_matches_for_this_value)} matches")
            
            # Combine all road conditions with OR logic - any road match counts as success
            combined_road_condition = all_road_conditions[0]
            for condition in all_road_conditions[1:]:
                combined_road_condition = combined_road_condition | condition
            
            road_filtered = state_matches[combined_road_condition]
            
            if not road_filtered.empty:
                road_matches = road_filtered
                print(f"    ✅ Road filtering successful: {len(road_matches)} total matches found")
                
                # Show which columns had matches for debugging
                matching_details = []
                for _, match_row in road_matches.iterrows():
                    matched_cols = []
                    for road_label, road_value in ocr_road_values:
                        if pd.notna(match_row['Highway']) and road_value.lower() in str(match_row['Highway']).lower():
                            matched_cols.append('Highway')
                        if pd.notna(match_row['Exit']) and road_value.lower() in str(match_row['Exit']).lower():
                            matched_cols.append('Exit')
                        if pd.notna(match_row['Street Address']) and road_value.lower() in str(match_row['Street Address']).lower():
                            matched_cols.append('Street Address')
                        if pd.notna(match_row['Road Name']) and road_value.lower() in str(match_row['Road Name']).lower():
                            matched_cols.append('Road Name')
                    if matched_cols:
                        matching_details.append(f"({', '.join(set(matched_cols))})")
                
                if matching_details:
                    print(f"      🎯 Matches found in columns: {', '.join(set(matching_details))}")
            else:
                print(f"    ⚠️ No road matches found, proceeding with all {len(road_matches)} state matches")
        else:
            print(f"  ⚠️ No road information available, using all state matches: {len(road_matches)}")
        
        # Step 7: Check name matches
        ocr_label = row['OCR_label']
        if pd.isna(ocr_label):
            ocr_words = []
        else:
            ocr_words = filter_meaningful_words(ocr_label)
        
        name_match_indices = []
        
        # If we have meaningful words to match, use name matching
        if ocr_words:
            print(f"  🔍 Name matching with words: {ocr_words}")
            
            for road_idx, road_row in road_matches.iterrows():
                chain_words = filter_meaningful_words(road_row['Chain'])
                name_words = filter_meaningful_words(road_row['name'])
                business_words = chain_words + name_words
                
                match_found = False
                matched_words = []
                
                for ocr_word in ocr_words:
                    for business_word in business_words:
                        if (ocr_word in business_word or business_word in ocr_word or
                            ocr_word == business_word):
                            match_found = True
                            matched_words.append(f"{ocr_word}→{business_word}")
                            break
                    if match_found:
                        break
                
                if match_found:
                    name_match_indices.append(road_idx)
                    print(f"    🎯 Match found: {matched_words}")
        else:
            # If no meaningful OCR label words, we'll flag all road matches for manual verification
            print(f"  ⚠️ No meaningful OCR label words, will flag all road matches for verification")
            name_match_indices = list(road_matches.index)
        
        print(f"  📊 Total potential matches: {len(name_match_indices)}")
        
        # Handle different match scenarios
        if len(name_match_indices) == 0:
            # No matches found
            result_df.loc[idx, 'Flagged'] = True
            result_df.loc[idx, 'Flag_Reason'] = "No Name/ZIP/State/Road matches found for empty address type"
            print(f"  ❌ No matches found for OCR label: '{row['OCR_label']}'")
            
        else:
            # We have matches - create rows for each match
            print(f"  ✅ Found {len(name_match_indices)} match(es) for OCR label: '{row['OCR_label']}'")
            print(f"  📊 Creating {len(name_match_indices)} row(s) for {len(name_match_indices)} match(es)")
            
            for match_num, match_idx in enumerate(name_match_indices):
                match_row = road_matches.loc[match_idx]
                
                if match_num == 0:
                    # First match: update the original row
                    target_idx = idx
                    result_df.loc[target_idx, 'Flagged'] = False
                    if len(name_match_indices) == 1:
                        result_df.loc[target_idx, 'Flag_Reason'] = "Single Name/ZIP/State/Road Match Found (empty address type)"
                        result_df.loc[target_idx, 'Manually Verified?'] = "Please Verify This"
                    else:
                        result_df.loc[target_idx, 'Flag_Reason'] = f"Multiple Name/ZIP/State/Road Matches Found (empty address type) - Match {match_num + 1} of {len(name_match_indices)}"
                        result_df.loc[target_idx, 'Manually Verified?'] = "Please Verify This"
                    
                    print(f"    📝 Filling original row with match {match_num + 1}")
                else:
                    # Additional matches: create new rows
                    new_row = result_df.loc[idx].copy()
                    new_row['Flagged'] = False
                    new_row['Flag_Reason'] = f"Multiple Name/ZIP/State/Road Matches Found (empty address type) - Match {match_num + 1} of {len(name_match_indices)}"
                    new_row['Manually Verified?'] = "Please Verify This"
                    
                    # Clear existing scraped data
                    for col in new_row.index:
                        if col.startswith('Scraped_'):
                            new_row[col] = None
                    
                    new_rows_to_add.append(new_row)
                    print(f"    ➕ Creating new row for match {match_num + 1}")
                
                # Fill scraped data for this match
                scraped_data = map_df2_to_scraped(match_row)
                
                if match_num == 0:
                    # Update original row
                    for col, value in scraped_data.items():
                        result_df.loc[idx, col] = value
                else:
                    # Update the new row we just created
                    for col, value in scraped_data.items():
                        new_rows_to_add[-1][col] = value
                
                print(f"      ✓ Filled data for {match_row['Chain']} - {match_row['name']}")
                print(f"        📝 OCR Label: '{row['OCR_label']}' → Matched Business: '{match_row['Chain']} - {match_row['name']}'")
    
    # Add all new rows to the result dataframe
    if new_rows_to_add:
        new_rows_df = pd.DataFrame(new_rows_to_add)
        result_df = pd.concat([result_df, new_rows_df], ignore_index=True)
        print(f"\n🎉 Added {len(new_rows_to_add)} new rows for multiple matches!")
    
    print(f"\n📊 Final result: {len(result_df)} rows (original: {len(df)}, added: {len(new_rows_to_add)})")
    return result_df

# Execute the matching implementation for "empty" address types
print("Starting matching logic for OCR_Address_Type=='empty' rows...")
final_result_empty = comprehensive_matching_logic_for_empty()
```
