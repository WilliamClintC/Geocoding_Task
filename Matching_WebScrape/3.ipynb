```python
# Select 10 random rows from df1 to process
import numpy as np
np.random.seed(45)  # For reproducible results
# Get 10 random row indices
random_rows = np.random.choice(df1.index, size=20, replace=False)
print(f"Selected 10 random rows from df1: {sorted(random_rows)}")

# Find phone column
phone_col = None
if 'Phone' in df1.columns:
    phone_col = 'Phone'
elif 'phone' in df1.columns:
    phone_col = 'phone'
else:
    phone_cols = [col for col in df1.columns if 'phone' in col.lower()]
    if phone_cols:
        phone_col = phone_cols[0]

if phone_col:
    print(f"Using phone column: {phone_col}")
else:
    print("No phone column found in df1")

print(f"\nPhone numbers from selected rows:")
for idx in sorted(random_rows):
    phone = df1.loc[idx, phone_col] if phone_col else "No phone column"
    print(f"Row {idx}: {phone}")
```

```python
# Search for phone number matches for all selected rows
search_columns = [col for col in df2.columns if col.lower() in ['phone', 'phone 2', 'fax']]
print(f"Searching in df2 columns: {search_columns}")

# Store all matches for all rows
all_matches = {}

for row_idx in random_rows:
    if phone_col and pd.notna(df1.loc[row_idx, phone_col]):
        phone_number = df1.loc[row_idx, phone_col]
        phone_str = str(phone_number)
        
        # Search for matches in df2
        row_matches = {}
        all_df2_matches = set()  # Track all df2 indices that match this df1 row
        
        for col in search_columns:
            exact_matches = df2[df2[col].astype(str) == phone_str]
            partial_matches = df2[df2[col].astype(str).str.contains(phone_str, case=False, na=False)]
            
            row_matches[col] = {
                'exact': len(exact_matches),
                'partial': len(partial_matches),
                'exact_indices': exact_matches.index.tolist(),
                'partial_indices': partial_matches.index.tolist()
            }
            
            # Collect all df2 matches for this df1 row
            all_df2_matches.update(exact_matches.index.tolist())
            all_df2_matches.update(partial_matches.index.tolist())
        
        # Calculate totals for this row
        total_exact = sum([row_matches[col]['exact'] for col in row_matches])
        total_partial = sum([row_matches[col]['partial'] for col in row_matches])
        
        all_matches[row_idx] = {
            'phone_number': phone_number,
            'matches': row_matches,
            'total_exact': total_exact,
            'total_partial': total_partial,
            'all_df2_indices': sorted(list(all_df2_matches))
        }
        
        # Enhanced output showing both df1 and df2 indices
        df2_indices_str = ', '.join(map(str, sorted(all_df2_matches))) if all_df2_matches else "None"
        print(f"df1 row {row_idx} ({phone_number}): {total_exact} exact, {total_partial} partial matches")
        print(f"   Matched df2 rows: [{df2_indices_str}]")
        
        # Show detailed breakdown by column if there are matches
        if all_df2_matches:
            for col in search_columns:
                if row_matches[col]['exact'] > 0 or row_matches[col]['partial'] > 0:
                    exact_list = ', '.join(map(str, row_matches[col]['exact_indices'])) if row_matches[col]['exact_indices'] else ""
                    partial_list = ', '.join(map(str, row_matches[col]['partial_indices'])) if row_matches[col]['partial_indices'] else ""
                    
                    if exact_list and partial_list:
                        print(f"    {col}: exact [{exact_list}], partial [{partial_list}]")
                    elif exact_list:
                        print(f"    {col}: exact [{exact_list}]")
                    elif partial_list:
                        print(f"    {col}: partial [{partial_list}]")
        
        print()  # Add blank line between rows
        
    else:
        all_matches[row_idx] = {
            'phone_number': None,
            'matches': {},
            'total_exact': 0,
            'total_partial': 0,
            'all_df2_indices': []
        }
        print(f"df1 row {row_idx}: No valid phone number")
        print(f"   Matched df2 rows: [None]")
        print()

print(f"Processed {len(all_matches)} df1 rows")

# Summary of all matches
total_df1_with_matches = sum(1 for match in all_matches.values() if match['all_df2_indices'])
total_unique_df2_matches = len(set().union(*[match['all_df2_indices'] for match in all_matches.values()]))

print(f"\nOverall Summary:")
print(f"- df1 rows with matches: {total_df1_with_matches}/{len(all_matches)}")
print(f"- Unique df2 rows involved: {total_unique_df2_matches}")
```

```python
print("=" * 60)
print("CREATING SINGLE COMBINED_DF WITH MATCH_ACCURACY COLUMN")
print("=" * 60)

# Create combined DataFrame for all selected rows
all_combined_rows = []

# Prepare renamed DataFrames
df1_full_renamed = df1.copy()
df1_full_renamed.columns = ['OCR_' + col for col in df1_full_renamed.columns]

df2_renamed = df2.copy()
df2_renamed.columns = ['Scraped_' + col for col in df2_renamed.columns]

print("Creating combined_df for all selected rows...")

# Process each row
for row_idx in random_rows:
    match_data = all_matches[row_idx]
    phone_number = match_data['phone_number']
    
    # Get all unique matching df2 indices for this row
    match_accuracy_map = {}
    if match_data['matches']:
        for col, results in match_data['matches'].items():
            # Process exact matches
            for df2_idx in results['exact_indices']:
                if df2_idx not in match_accuracy_map:
                    match_accuracy_map[df2_idx] = []
                match_accuracy_map[df2_idx].append(f"exact phone number ({col})")
            
            # Process partial matches (if not already exact)
            for df2_idx in results['partial_indices']:
                if df2_idx not in match_accuracy_map:
                    match_accuracy_map[df2_idx] = []
                if df2_idx not in results['exact_indices']:
                    match_accuracy_map[df2_idx].append(f"partial phone number ({col})")
    
    total_matches = len(match_accuracy_map)
    
    # Determine flagged status
    flag_reason = None
    if total_matches == 0:
        flag_reason = "No matches found"
    elif total_matches > 1:
        flag_reason = "Multiple matches found"
    
    flagged = flag_reason is not None
    
    # Create combined rows for this df1 row
    if total_matches > 0:
        # Create rows for each match
        for df2_idx, match_types in sorted(match_accuracy_map.items()):
            df1_row = df1_full_renamed.loc[row_idx].to_dict()  # Use .loc instead of .iloc
            df2_row = df2_renamed.iloc[df2_idx].to_dict()      # .iloc is fine here since df2_idx is positional
            
            combined_row = {**df1_row, **df2_row}
            combined_row['match_accuracy'] = ', '.join(match_types)
            combined_row['df1_source_row'] = row_idx
            combined_row['df2_matched_row'] = df2_idx
            combined_row['Flagged'] = flagged
            combined_row['Flag_Reason'] = flag_reason
            
            all_combined_rows.append(combined_row)
    else:
        # No matches - create row with df1 data and empty df2 data
        df1_row = df1_full_renamed.loc[row_idx].to_dict()  # Use .loc instead of .iloc
        empty_df2_row = {col: None for col in df2_renamed.columns}
        
        combined_row = {**df1_row, **empty_df2_row}
        combined_row['match_accuracy'] = "No matches"
        combined_row['df1_source_row'] = row_idx
        combined_row['df2_matched_row'] = None
        combined_row['Flagged'] = flagged
        combined_row['Flag_Reason'] = flag_reason
        
        all_combined_rows.append(combined_row)

# Create final DataFrame
combined_df = pd.DataFrame(all_combined_rows)

print(f"Combined DataFrame created: {combined_df.shape[0]} rows, {combined_df.shape[1]} columns")

# Summary statistics
flagged_count = len(combined_df[combined_df['Flagged'] == True])
unflagged_count = len(combined_df[combined_df['Flagged'] == False])

print(f"\nSummary:")
print(f"- Total rows processed from df1: {len(random_rows)}")
print(f"- Total rows in combined_df: {len(combined_df)}")
print(f"- Flagged rows: {flagged_count}")
print(f"- Unflagged rows: {unflagged_count}")

# Show flag reasons
print(f"\nFlag reason breakdown:")
flag_summary = combined_df['Flag_Reason'].value_counts(dropna=False)
for reason, count in flag_summary.items():
    print(f"- {reason if pd.notna(reason) else 'No flag'}: {count} rows")

# Show key columns
print(f"\nFirst 10 rows of key columns:")
key_cols = ['df1_source_row', 'df2_matched_row', 'match_accuracy', 'Flagged', 'Flag_Reason']
print(combined_df[key_cols].head(10).to_string(index=False))
```

```python
# Analyze flagging in detail
print("FLAGGING ANALYSIS")
print("=" * 50)

# Show flag reason breakdown
print("Flag reason summary:")
flag_summary = combined_df['Flag_Reason'].value_counts(dropna=False)
for reason, count in flag_summary.items():
    print(f"- {reason if pd.notna(reason) else 'No flag'}: {count} rows")

print("\n" + "=" * 50)
print("DETAILED FLAGGING EXAMPLES")
print("=" * 50)

# Show examples of each flag type
for reason in flag_summary.index:
    if pd.notna(reason):  # Only show actual flag reasons
        print(f"\n--- {reason} ---")
        examples = combined_df[combined_df['Flag_Reason'] == reason]
        
        # Group by df1_source_row to show which df1 rows have this issue
        df1_rows_with_issue = examples['df1_source_row'].unique()
        print(f"df1 rows affected: {sorted(df1_rows_with_issue)}")
        
        # Show details for each df1 row
        for df1_row in sorted(df1_rows_with_issue):
            df1_examples = examples[examples['df1_source_row'] == df1_row]
            df2_matches = df1_examples['df2_matched_row'].tolist()
            phone_val = df1_examples['OCR_Phone'].iloc[0] if 'OCR_Phone' in df1_examples.columns else "N/A"
            
            print(f"  df1 row {df1_row} (phone: {phone_val}):")
            print(f"     matches df2 rows: {df2_matches}")
            
            # Show match accuracy for each
            for _, row in df1_examples.iterrows():
                accuracy = row['match_accuracy']
                df2_row = row['df2_matched_row']
                print(f"      df2 row {df2_row}: {accuracy}")
        
        if len(examples) > 5:
            print(f"  ... showing first few examples of {len(examples)} total rows")

print("\n" + "=" * 50)
print("SUMMARY OF THE FLAGGING LOGIC")
print("=" * 50)
print("The code flags records when:")
print("1. 'No matches found' - A df1 row has no matching phone numbers in df2")
print("2. 'Multiple matches found' - A df1 row matches MORE THAN ONE row in df2")
print("")
print("When something 'exists in two rows', it means:")
print("- One df1 phone number was found in multiple df2 rows")
print("- This creates multiple combined rows (one for each df2 match)")
print("- ALL of these rows get flagged as 'Multiple matches found'")
print("- This helps identify potential duplicates or data quality issues")
```
