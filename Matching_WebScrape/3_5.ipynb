```python
# Create a new column called "Address_Type" - initially empty
df['Address_Type'] = 'empty'

# Set "Address_Type" to "Exit" if OCR_address_standardized_OFF_parenthesis contains "Exit", "Ex", or "X" patterns
# This will catch patterns like "Exit", "I-15 Exit 10", "I-710 Ex 1 B SB", "Everett Tpke X 10", "Garden State Pkwy X 157", "I-40 xit 325", etc.
# Now recognizes any form of "xit" including accented characters
exit_pattern = r'([Ee]?xit|Ex\s+\d+|Ex$|\bEx\b|\bX\s+\d+)'
df.loc[df['OCR_address_standardized_OFF_parenthesis'].str.contains(exit_pattern, case=False, na=False, regex=True), 'Address_Type'] = 'Exit'

#display OCR_address_standardized_OFF_parenthesis OCR_address_standardized_ON_parenthesis Address_Type
df[['OCR_address_standardized_OFF_parenthesis', 'OCR_address_standardized_ON_parenthesis', 'Address_Type','OCR_zip_code']]
```

```python
def remove_zipcode_from_all_rows(df):
    """
    Remove zip codes from address strings for all rows in the DataFrame.
    Handles ZIP codes both with and without leading zeros (e.g., both "1105" and "01105").
    
    Args:
        df: DataFrame containing the data
    
    Returns:
        DataFrame with updated addresses
    """
    rows_processed = 0
    rows_updated = 0
    
    for index in range(len(df)):
        zip_code = str(df.iloc[index]['OCR_zip_code'])
        current_address = str(df.iloc[index]['OCR_address_standardized_OFF_parenthesis'])
        
        if zip_code != 'nan' and zip_code.strip():
            # Clean the zip code and ensure it's numeric
            zip_code_clean = zip_code.strip()
            
            # Create both versions of the ZIP code (with and without leading zero)
            zip_codes_to_remove = set()
            
            if zip_code_clean.isdigit():
                # Add the original ZIP code
                zip_codes_to_remove.add(zip_code_clean)
                
                # If it's 4 digits, add the 5-digit version with leading zero
                if len(zip_code_clean) == 4:
                    zip_codes_to_remove.add('0' + zip_code_clean)
                
                # If it's 5 digits and starts with 0, add the version without leading zero
                elif len(zip_code_clean) == 5 and zip_code_clean.startswith('0'):
                    zip_codes_to_remove.add(zip_code_clean[1:])
            
            # Check if any version of the ZIP code exists in the address
            address_updated = False
            updated_address = current_address
            
            for zip_to_remove in zip_codes_to_remove:
                if zip_to_remove in updated_address:
                    # Remove the zip code from the address string
                    updated_address = updated_address.replace(zip_to_remove, '').strip()
                    address_updated = True
            
            if address_updated:
                # Clean up multiple spaces and trailing/leading commas
                updated_address = ' '.join(updated_address.split())  # Remove multiple spaces
                updated_address = updated_address.replace(' ,', ',')  # Fix space before comma
                updated_address = updated_address.replace(', ,', ',')  # Fix double commas
                updated_address = updated_address.strip(' ,')  # Remove leading/trailing spaces and commas
                
                # Update the DataFrame
                df.iloc[index, df.columns.get_loc('OCR_address_standardized_OFF_parenthesis')] = updated_address
                rows_updated += 1
        
        rows_processed += 1
    
    print(f"Processed {rows_processed} rows, updated {rows_updated} addresses")
    return df

# If you want to apply this to the entire DataFrame, uncomment the line below:
# df = remove_zipcode_from_all_rows(df)

print("Enhanced ZIP code removal function is ready to use!")
print("\nFeatures:")
print("- Handles ZIP codes with and without leading zeros")
print("- Example: If OCR_zip_code is '1105', it will remove both '1105' and '01105' from addresses")
print("- Example: If OCR_zip_code is '01105', it will remove both '01105' and '1105' from addresses")
print("\nUsage:")
print("df = remove_zipcode_from_all_rows(df)")
print("\nNote: Make sure to save the DataFrame after processing if you want to keep the changes.")
```

```python
# Clean OCR_address_standardized_OFF_parenthesis column by removing leading "-" characters
def clean_leading_dashes(df):
    """
    Remove leading "-" characters from OCR_address_standardized_OFF_parenthesis column
    Example: "- 43859 N Sierra " becomes "43859 N Sierra "
    """
    print("Cleaning leading dashes from OCR_address_standardized_OFF_parenthesis column...")
    
    # Count how many entries have leading dashes
    leading_dash_count = df['OCR_address_standardized_OFF_parenthesis'].str.startswith('- ', na=False).sum()
    print(f"Found {leading_dash_count} entries with leading dashes")
    
    # Remove leading "- " from the addresses
    df['OCR_address_standardized_OFF_parenthesis'] = df['OCR_address_standardized_OFF_parenthesis'].str.replace(r'^- ', '', regex=True)
    
    print("Cleaning completed!")
    return df

# Apply the cleaning function
df = clean_leading_dashes(df)

# Display sample of cleaned data
print("\nSample of cleaned data:")
df[['OCR_address_standardized_OFF_parenthesis']]
```
