```python
import os
import requests
from bs4 import BeautifulSoup
import time
from urllib.parse import urlparse

def extract_image_filename_from_div(target_div):
    """
    Extract image filename from img src within the target div
    Example: from "https://www.truckstopsandservices.com/use_images/ads/talogo.jpg"
    extract "talogo"
    """
    if not target_div:
        return None
    
    # Find img tag within the div
    img_tag = target_div.find('img')
    if not img_tag:
        return None
    
    # Get the src attribute
    img_src = img_tag.get('src')
    if not img_src:
        return None
    
    # Extract filename from URL
    parsed_url = urlparse(img_src)
    filename_with_ext = os.path.basename(parsed_url.path)
    
    # Remove file extension to get just the name
    filename_without_ext = os.path.splitext(filename_with_ext)[0]
    
    return filename_without_ext
    """
    Process catscale3 entries and extract image filenames from specific div:
    1. Find entries where Chain == "catscale3"
    2. Open their full_url
    3. Extract image filename from <div class="pop_ad_holder popb_nonrotate">
    4. Update Chain to original + image_filename
    """
    catscale3_indices = df2[df2["Chain"] == "catscale3"].index.tolist()
    processed_count = 0
    image_found_count = 0
    
    print(f"Processing {len(catscale3_indices)} catscale3 entries...")
    
    for idx in catscale3_indices:
        try:
            url = df2.loc[idx, "full_url"]
            
            # Make request to the URL
            headers = {
                'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
            }
            
            response = requests.get(url, headers=headers, timeout=10)
            response.raise_for_status()
            
            # Parse HTML
            soup = BeautifulSoup(response.content, 'html.parser')
            
            # Find the specific div
            target_div = soup.find('div', class_='pop_ad_holder popb_nonrotate')
            
            if target_div:
                # Extract image filename
                image_filename = extract_image_filename_from_div(target_div)
                
                if image_filename:
                    # Update the chain name with the extracted filename
                    original_chain = df2.loc[idx, "Chain"]
                    new_chain = f"{original_chain} + {image_filename}"
                    df2.loc[idx, "Chain"] = new_chain
                    image_found_count += 1
            
            processed_count += 1
            
            # Add delay to be respectful to the server
            time.sleep(0.3)
                
        except Exception as e:
            processed_count += 1
            continue
    
    print(f"Processed: {processed_count} entries")
    print(f"Found images in: {image_found_count} entries")
    
    return processed_count, image_found_count
```

```python
def process_all_catscale3_with_image_extraction():
    """
    Process ALL remaining catscale3 entries and extract image filenames
    """
    # Find remaining unprocessed catscale3 entries
    remaining_catscale3 = df2[df2["Chain"] == "catscale3"]
    catscale3_indices = remaining_catscale3.index.tolist()
    
    processed_count = 0
    image_found_count = 0
    errors = 0
    
    print(f"Starting to process ALL {len(catscale3_indices)} remaining catscale3 entries...")
    print("Extracting image filenames from div...")
    
    for i, idx in enumerate(catscale3_indices):
        try:
            url = df2.loc[idx, "full_url"]
            
            # Progress indicator
            if i % 25 == 0:
                print(f"\nProgress: {i}/{len(catscale3_indices)} ({i/len(catscale3_indices)*100:.1f}%)")
            
            # Make request to the URL
            headers = {
                'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
            }
            
            response = requests.get(url, headers=headers, timeout=10)
            response.raise_for_status()
            
            # Parse HTML
            soup = BeautifulSoup(response.content, 'html.parser')
            
            # Find the specific div
            target_div = soup.find('div', class_='pop_ad_holder popb_nonrotate')
            
            if target_div:
                # Extract image filename
                image_filename = extract_image_filename_from_div(target_div)
                
                if image_filename:
                    # Update the chain name with the extracted filename
                    original_chain = df2.loc[idx, "Chain"]
                    new_chain = f"{original_chain} + {image_filename}"
                    df2.loc[idx, "Chain"] = new_chain
                    image_found_count += 1
                    
                    if i % 25 == 0:  # Show sample updates
                        print(f"  Sample update: {original_chain} → {new_chain}")
            
            processed_count += 1
            
            # Add delay to be respectful to the server
            time.sleep(0.3)
                
        except Exception as e:
            errors += 1
            processed_count += 1
            if errors % 10 == 0:  # Report every 10 errors
                print(f"  Error count: {errors}")
            continue
    
    print(f"\n=== Final Summary ===")
    print(f"Processed: {processed_count} entries")
    print(f"Found images in: {image_found_count} entries")
    print(f"Errors: {errors}")
    
    # Show final distribution
    print(f"\nFinal catscale3 distribution:")
    final_catscale_variants = df2[df2["Chain"].str.contains("catscale3", na=False)]
    final_counts = final_catscale_variants["Chain"].value_counts()
    print(final_counts.head(10))
    
    return processed_count, image_found_count, errors

print("Updated function created!")
print("The function now extracts image filenames (like 'talogo') from img src attributes")
print("instead of searching for 'ta-petro' text.")
print("\nTo process all remaining entries, run:")
print("processed_all, found_all, errors_all = process_all_catscale3_with_image_extraction()")
print(f"\nCurrently {len(df2[df2['Chain'] == 'catscale3'])} entries remaining to process.")

# Run the processing function to extract image filenames from catscale3 entries
processed_count, image_found_count, errors = process_all_catscale3_with_image_extraction()

# Show final results
catscale3_variants = df2[df2["Chain"].str.contains("catscale3", na=False)]
print(f"\nFinal catscale3 distribution:")
print(catscale3_variants["Chain"].value_counts())
```
