```python
# COMPREHENSIVE EXTRACTION FUNCTION - TAILORED TO YOUR ACTUAL DATA (PRESERVING EXISTING VALUES)
import re
import pandas as pd

def extract_road_and_exit_info(row):
    """
    Extract secondary and tertiary road and exit information from complex address strings,
    specifically tailored to the patterns observed in the actual data.
    
    This version preserves any existing values in the relevant columns.
    
    Parameters:
        row: DataFrame row containing address_standardized_OFF_parenthesis and possibly existing values
        
    Returns:
        Dictionary with Exit_Number_2, Exit_Number_3, Secondary_Road, and Tertiary_Road.
    """
    # Initialize with existing values from the row (or None if they don't exist)
    result = {
        'Exit_Number_2': row.get('Exit_Number_2'), 
        'Exit_Number_3': row.get('Exit_Number_3'),
        'Secondary_Road': row.get('Secondary_Road'),
        'Tertiary_Road': row.get('Tertiary_Road')
    }
    
    # Get the address string
    address_string = row['address_standardized_OFF_parenthesis']
    
    # Skip processing if no address or no slash
    if pd.isna(address_string) or not isinstance(address_string, str) or '/' not in address_string:
        return result
    
    # Split the address by '/'
    parts = [part.strip() for part in address_string.split('/')]
    
    # Process second part (if exists)
    if len(parts) > 1:
        second_part = parts[1]
        
        # More flexible road pattern - look for common highway prefixes
        if result['Secondary_Road'] is None:  # Only update if no existing value
            road_match = re.search(r'\b(I|US|SR|UT|CA|NV|AZ|NM)[-\s]?(\d+)', second_part)
            if road_match:
                prefix = road_match.group(1)
                number = road_match.group(2)
                result['Secondary_Road'] = f"{prefix}-{number}"
        
        if result['Exit_Number_2'] is None:  # Only update if no existing value
            # Try to match "Exit 123", "X 123", or "Xt 123" pattern first
            exit_match = re.search(r'[XE](?:xit|t)?\s*(\d+)', second_part)
            if exit_match:
                result['Exit_Number_2'] = exit_match.group(1)
            else:
                # If no explicit exit, just look for any number
                num_match = re.search(r'\b(\d+)\b', second_part)
                if num_match:
                    result['Exit_Number_2'] = num_match.group(1)
    
    # Process third part (if exists)
    if len(parts) > 2:
        third_part = parts[2]
        
        # More flexible road pattern
        if result['Tertiary_Road'] is None:  # Only update if no existing value
            road_match = re.search(r'\b(I|US|SR|UT|CA|NV|AZ|NM)[-\s]?(\d+)', third_part)
            if road_match:
                prefix = road_match.group(1)
                number = road_match.group(2)
                result['Tertiary_Road'] = f"{prefix}-{number}"
        
        if result['Exit_Number_3'] is None:  # Only update if no existing value
            # Try to match exit pattern first
            exit_match = re.search(r'[XE](?:xit|t)?\s*(\d+)', third_part)
            if exit_match:
                result['Exit_Number_3'] = exit_match.group(1)
            else:
                # If no explicit exit, look for any number
                num_match = re.search(r'\b(\d+)\b', third_part)
                if num_match:
                    result['Exit_Number_3'] = num_match.group(1)
    
    return result

print("EXTRACTING ROAD AND EXIT INFORMATION FROM ADDRESSES:")
print("---------------------------------------------------")

# Test with an example to demonstrate preservation of existing values
example_row = {
    'address_standardized_OFF_parenthesis': "I-15 X 305 B / I-80 WB X 122 / UT 201 EB X17",
    'Exit_Number_2': '999',  # Existing value that should be preserved
    'Secondary_Road': None,
    'Exit_Number_3': None,
    'Tertiary_Road': 'EXISTING-ROAD'  # Another existing value
}

print("Example with existing data:")
print(f"  Address: '{example_row['address_standardized_OFF_parenthesis']}'")
print(f"  Existing Exit_Number_2: {example_row['Exit_Number_2']}")
print(f"  Existing Tertiary_Road: {example_row['Tertiary_Road']}")

result = extract_road_and_exit_info(example_row)
print("\nResults after extraction (preserving existing values):")
for key, value in result.items():
    print(f"  {key}: {value}")

# Apply to the dataframe
print("\nApplying extraction function to the entire dataframe...")

# First print summary of existing data
print("\nBefore extraction - current data:")
print(f"Rows with Exit_Number_2: {df1['Exit_Number_2'].notna().sum() if 'Exit_Number_2' in df1.columns else 0}")
print(f"Rows with Secondary_Road: {df1['Secondary_Road'].notna().sum() if 'Secondary_Road' in df1.columns else 0}")
print(f"Rows with Exit_Number_3: {df1['Exit_Number_3'].notna().sum() if 'Exit_Number_3' in df1.columns else 0}")
print(f"Rows with Tertiary_Road: {df1['Tertiary_Road'].notna().sum() if 'Tertiary_Road' in df1.columns else 0}")

# Apply the function row by row to preserve existing values
results = df1.apply(extract_road_and_exit_info, axis=1)
extracted_df = pd.DataFrame(results.tolist(), index=df1.index)

# Ensure all necessary columns exist in df1
for col in extracted_df.columns:
    if col not in df1.columns:
        df1[col] = None

# Update df1 using masks - much safer than iterating through indices
for col in extracted_df.columns:
    mask = extracted_df[col].notna()
    if mask.any():
        df1.loc[mask, col] = extracted_df.loc[mask, col]

# Display summary of extracted data
print("\nAfter extraction (preserving existing values):")
print(f"Rows with Exit_Number_2: {df1['Exit_Number_2'].notna().sum()}")
print(f"Rows with Secondary_Road: {df1['Secondary_Road'].notna().sum()}")
print(f"Rows with Exit_Number_3: {df1['Exit_Number_3'].notna().sum()}")
print(f"Rows with Tertiary_Road: {df1['Tertiary_Road'].notna().sum()}")

# Display examples with extracted information
sample_df = df1[(df1['Exit_Number_2'].notna()) | (df1['Secondary_Road'].notna()) | 
               (df1['Exit_Number_3'].notna()) | (df1['Tertiary_Road'].notna())]

if len(sample_df) > 0:
    print(f"\nFound {len(sample_df)} rows with extracted information. Here are some examples:")
    display(sample_df[['address_standardized_OFF_parenthesis', 'Secondary_Road', 'Exit_Number_2', 
                      'Tertiary_Road', 'Exit_Number_3']].head(5))
else:
    print("\nNo rows with extracted information found.")
```

```python
# IMPROVED EXTRACTION FUNCTION - PRESERVE EXISTING DATA
import re
import pandas as pd

def extract_info_preserving_existing_data(address_string, existing_exit_number_2=None, existing_secondary_road=None, 
                                         existing_exit_number_3=None, existing_tertiary_road=None):
    """
    Extract information from addresses based on the actual patterns in the dataframe,
    while preserving any existing data in the columns.
    
    Parameters:
        address_string (str): The address string to parse
        existing_exit_number_2: Existing value for Exit_Number_2, if any
        existing_secondary_road: Existing value for Secondary_Road, if any
        existing_exit_number_3: Existing value for Exit_Number_3, if any
        existing_tertiary_road: Existing value for Tertiary_Road, if any
        
    Returns:
        dict: Dictionary with Exit_Number_2, Secondary_Road, etc.
    """
    # Start with existing values
    result = {
        'Exit_Number_2': existing_exit_number_2, 
        'Exit_Number_3': existing_exit_number_3,
        'Secondary_Road': existing_secondary_road,
        'Tertiary_Road': existing_tertiary_road
    }
    
    # If no address or no slash, just return existing values
    if pd.isna(address_string) or not isinstance(address_string, str) or '/' not in address_string:
        return result
    
    # Split the address by '/'
    parts = [part.strip() for part in address_string.split('/')]
    
    # Process second part (if exists) and only update if we don't have existing values
    if len(parts) > 1:
        second_part = parts[1]
        
        # Look for road pattern - very flexible now
        road_match = re.search(r'\b(I|US|SR|UT|CA|NV|AZ|NM)[-\s]?(\d+)', second_part)
        if road_match and result['Secondary_Road'] is None:  # Only update if no existing value
            prefix = road_match.group(1)
            number = road_match.group(2)
            result['Secondary_Road'] = f"{prefix}-{number}"
            
        # Look for exit number patterns including "Xt"
        if result['Exit_Number_2'] is None:  # Only update if no existing value
            # First check for exit pattern including Xt
            exit_match = re.search(r'[XE](?:xit|t)?\s*(\d+)', second_part)
            if exit_match:
                result['Exit_Number_2'] = exit_match.group(1)
            else:
                # If no explicit exit, just look for any number
                num_match = re.search(r'\b(\d+)\b', second_part)
                if num_match:
                    result['Exit_Number_2'] = num_match.group(1)
    
    # Process third part (if exists) and only update if we don't have existing values
    if len(parts) > 2:
        third_part = parts[2]
        
        # Look for road pattern
        road_match = re.search(r'\b(I|US|SR|UT|CA|NV|AZ|NM)[-\s]?(\d+)', third_part)
        if road_match and result['Tertiary_Road'] is None:  # Only update if no existing value
            prefix = road_match.group(1)
            number = road_match.group(2)
            result['Tertiary_Road'] = f"{prefix}-{number}"
            
        # Look for exit number patterns including "Xt"
        if result['Exit_Number_3'] is None:  # Only update if no existing value
            # First check for exit pattern
            exit_match = re.search(r'[XE](?:xit|t)?\s*(\d+)', third_part)
            if exit_match:
                result['Exit_Number_3'] = exit_match.group(1)
            else:
                # If no explicit exit, look for any number
                num_match = re.search(r'\b(\d+)\b', third_part)
                if num_match:
                    result['Exit_Number_3'] = num_match.group(1)
    
    return result

# Function to apply extraction to a row
def extract_from_row(row):
    # Get the existing values (or None if they don't exist)
    existing_exit_number_2 = row.get('Exit_Number_2')
    existing_secondary_road = row.get('Secondary_Road')
    existing_exit_number_3 = row.get('Exit_Number_3')
    existing_tertiary_road = row.get('Tertiary_Road')
    
    # Extract information from the address while preserving existing values
    address = row['address_standardized_OFF_parenthesis']
    return extract_info_preserving_existing_data(
        address, 
        existing_exit_number_2, 
        existing_secondary_road, 
        existing_exit_number_3, 
        existing_tertiary_road
    )

print("Applying the extraction function while preserving existing data...")

# Apply the function row by row to preserve existing data
results = df1.apply(extract_from_row, axis=1)
extracted_df = pd.DataFrame(results.tolist(), index=df1.index)

# Ensure all necessary columns exist in df1
for col in extracted_df.columns:
    if col not in df1.columns:
        df1[col] = None

# Update df1 using a safer approach that preserves indices
for col in extracted_df.columns:
    # Only update where extracted_df has non-null values
    mask = extracted_df[col].notna()
    if mask.any():
        df1.loc[mask, col] = extracted_df.loc[mask, col]

# Display summary information
print("\nExtraction complete. Summary of extracted data:")
print(f"Rows with Exit_Number_2: {df1['Exit_Number_2'].notna().sum()}")
print(f"Rows with Secondary_Road: {df1['Secondary_Road'].notna().sum()}")
print(f"Rows with Exit_Number_3: {df1['Exit_Number_3'].notna().sum()}")
print(f"Rows with Tertiary_Road: {df1['Tertiary_Road'].notna().sum()}")

# Display examples with extracted information
sample_df = df1[(df1['Exit_Number_2'].notna()) | (df1['Secondary_Road'].notna()) | 
               (df1['Exit_Number_3'].notna()) | (df1['Tertiary_Road'].notna())]

if len(sample_df) > 0:
    print(f"\nFound {len(sample_df)} rows with extracted information. Here are some examples:")
    display(sample_df[['address_standardized_OFF_parenthesis', 'Secondary_Road', 'Exit_Number_2', 
                      'Tertiary_Road', 'Exit_Number_3']].head(15))
else:
    print("\nNo rows with extracted information found.")
```

```python
# FINAL PROCESSING - ADD SECONDARY_ROAD AND TERTIARY_ROAD (PRESERVING EXISTING DATA)
import re

# Function to extract Secondary_Road while preserving existing data
def extract_secondary_road(row):
    # If there's already a value, keep it
    if pd.notna(row.get('Secondary_Road')):
        return row['Secondary_Road']
        
    address = row['address_standardized_OFF_parenthesis']
    if pd.isna(address) or '/' not in address:
        return None
    
    parts = address.split('/')
    if len(parts) < 2:
        return None
    
    second_part = parts[1].strip()
    # Look for road pattern (I-80, UT 201, etc.)
    road_match = re.search(r'\b(I|US|SR|UT|CA|NV|AZ|NM)[-\s]?(\d+)', second_part)
    if road_match:
        prefix = road_match.group(1)
        number = road_match.group(2)
        return f"{prefix}-{number}"
    
    return None

# Function to extract Exit_Number_2 while preserving existing data
def extract_exit_number_2(row):
    # If there's already a value, keep it
    if pd.notna(row.get('Exit_Number_2')):
        return row['Exit_Number_2']
        
    address = row['address_standardized_OFF_parenthesis']
    if pd.isna(address) or '/' not in address:
        return None
    
    parts = address.split('/')
    if len(parts) < 2:
        return None
    
    second_part = parts[1].strip()
    # Look for exit pattern - now includes Xt as well
    exit_match = re.search(r'[XE](?:xit|t)?\s*(\d+)', second_part)
    if exit_match:
        return exit_match.group(1)
    
    # If no explicit exit marker, look for any number
    num_match = re.search(r'\b(\d+)\b', second_part)
    if num_match:
        return num_match.group(1)
    
    return None

# Function to extract Tertiary_Road while preserving existing data
def extract_tertiary_road(row):
    # If there's already a value, keep it
    if pd.notna(row.get('Tertiary_Road')):
        return row['Tertiary_Road']
        
    address = row['address_standardized_OFF_parenthesis']
    if pd.isna(address) or '/' not in address:
        return None
    
    parts = address.split('/')
    if len(parts) < 3:
        return None
    
    third_part = parts[2].strip()
    # Look for road pattern
    road_match = re.search(r'\b(I|US|SR|UT|CA|NV|AZ|NM)[-\s]?(\d+)', third_part)
    if road_match:
        prefix = road_match.group(1)
        number = road_match.group(2)
        return f"{prefix}-{number}"
    
    return None

# Apply these functions to the dataframe, preserving existing values
print("ADDING SECONDARY_ROAD AND TERTIARY_ROAD (PRESERVING EXISTING DATA):")
print("--------------------------------------------------------------")
print("Before extraction:")
print(f"Rows with Secondary_Road: {df1['Secondary_Road'].notna().sum()}")
print(f"Rows with Tertiary_Road: {df1['Tertiary_Road'].notna().sum() if 'Tertiary_Road' in df1.columns else 0}")

# Add a test case for our new pattern
print("\nTesting with new example pattern:")
test_example = {
    'address_standardized_OFF_parenthesis': "I-80 Bus / US 50 Xt 1"
}
print(f"Example: {test_example['address_standardized_OFF_parenthesis']}")
test_example['Secondary_Road'] = extract_secondary_road(test_example)
test_example['Exit_Number_2'] = extract_exit_number_2(test_example)
print(f"Extracted Secondary_Road: {test_example['Secondary_Road']}")
print(f"Extracted Exit_Number_2: {test_example['Exit_Number_2']}")

# Create temporary Series to hold new values
secondary_road_series = df1.apply(extract_secondary_road, axis=1)
tertiary_road_series = df1.apply(extract_tertiary_road, axis=1)

# Ensure the columns exist
if 'Secondary_Road' not in df1.columns:
    df1['Secondary_Road'] = None
if 'Tertiary_Road' not in df1.columns:
    df1['Tertiary_Road'] = None

# Update only where we have new values
secondary_mask = secondary_road_series.notna()
tertiary_mask = tertiary_road_series.notna()

if secondary_mask.any():
    df1.loc[secondary_mask, 'Secondary_Road'] = secondary_road_series[secondary_mask]
if tertiary_mask.any():
    df1.loc[tertiary_mask, 'Tertiary_Road'] = tertiary_road_series[tertiary_mask]

# FINAL STEP: Save the updated dataframe with all extracted columns to Add_1.csv
print("SAVING FINAL RESULTS:")
print("--------------------")
print(f"Total rows in dataframe: {len(df1)}")

# Final summary of extracted data
print("\nFinal extraction summary:")
print(f"Rows with Exit_Number_2: {df1['Exit_Number_2'].notna().sum()}")
print(f"Rows with Secondary_Road: {df1['Secondary_Road'].notna().sum()}")
print(f"Rows with Exit_Number_3: {df1['Exit_Number_3'].notna().sum()}")
print(f"Rows with Tertiary_Road: {df1['Tertiary_Road'].notna().sum()}")

# List all columns in the final dataframe
print("\nAll columns in the final dataframe:")
for i, col in enumerate(df1.columns):
    print(f"{i+1}. {col}")

# Save both dataframes
print("\nSaving the final dataframes...")
df1.to_csv('Add_1.csv', index=False)
df2.to_csv('Add_1_scraped.csv', index=False)

print("âœ… Successfully saved dataframes to:")
print("   1. Add_1.csv - Contains all the extracted road and exit information")
print("   2. Add_1_scraped.csv - Contains the scraped data")
print("\nâœ… The extraction preserves existing data in Exit_Number_2 and Secondary_Road columns")
print("   This means the notebook adds new entries without removing or replacing existing data")

# Verify the files were created
import os
if os.path.exists('Add_1.csv'):
    file_size = os.path.getsize('Add_1.csv') / 1024  # Size in KB
    print(f"\nAdd_1.csv file size: {file_size:.2f} KB")
if os.path.exists('Add_1_scraped.csv'):
    file_size = os.path.getsize('Add_1_scraped.csv') / 1024  # Size in KB
    print(f"Add_1_scraped.csv file size: {file_size:.2f} KB")
```
