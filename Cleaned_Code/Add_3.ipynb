# **Geocoding Task Notebook Explanation**

This notebook implements various matching algorithms between two datasets:
- `df1 - Main dataset with gas stations
- `df2- Scraped dataset with additional information

We'll match these datasets using several criteria:
1. Phone number matching
2. ZIP code matching
3. City name matching
4. Exit number matching
5. State matching
6. Road name matching
7. Chain name matching
8. Label text matching

Each matching will create a new column in df1 containing the row IDs of matching rows in df2.

```python
# Load the required datasets
import pandas as pd
import re
import numpy as np



# Display the first few rows of each dataset
print("Main dataset (df1) shape:", df1.shape)
print("Scraped dataset (df2) shape:", df2.shape)

# Display a sample of each dataset
display(df1.head(3))
display(df2.head(3))

# Reset the indices to ensure they're sequential
df1 = df1.reset_index(drop=True)
df2 = df2.reset_index(drop=True)
```

```python
# STEP 1: Match by phone number
# Create the new column to store matching row IDs
df1['phone_scraped_matches_row_ids'] = None

# Function to clean phone numbers by removing non-numeric characters
def clean_phone_number(phone_str):
    if pd.isna(phone_str):
        return ''
    # Extract only digits from the phone number
    digits_only = re.sub(r'\D', '', str(phone_str))
    # If we have at least 7 digits, consider it a valid phone number
    if len(digits_only) >= 7:
        return digits_only
    return ''

# Clean phone numbers in df1
df1_clean_phones = df1['phone'].apply(clean_phone_number)

# Initialize the phone columns to check in df2
phone_columns = ['Phone', 'Phone 2', 'Phone 3', 'Phone 4', 'Phone 5', 'Fax']

# Create cleaned versions of all phone columns in df2
df2_clean_phones = {}
for col in phone_columns:
    if col in df2.columns:
        df2_clean_phones[col] = df2[col].apply(clean_phone_number)
    else:
        print(f"Warning: Column '{col}' not found in df2")

# Match phone numbers and store row IDs
for i, phone in enumerate(df1_clean_phones):
    if phone:  # If phone number is not empty
        matches = []
        for col, clean_phones in df2_clean_phones.items():
            # Find all rows where this phone number matches
            matching_idx = clean_phones[clean_phones == phone].index.tolist()
            matches.extend(matching_idx)
        
        # Remove duplicates and store the matches
        if matches:
            df1.at[i, 'phone_scraped_matches_row_ids'] = str(list(set(matches)))

# Count how many rows got matches
matched_count = df1['phone_scraped_matches_row_ids'].notna().sum()
print(f"Phone number matching complete: {matched_count} rows in df1 have matches in df2")
```

```python
# STEP 2: Match by ZIP code
# Create the new column to store matching row IDs
df1['ZIP_scraped_matches_row_ids'] = None

# Function to clean ZIP codes by ensuring they're strings and removing extra spaces
def clean_zip_code(zip_str):
    if pd.isna(zip_str):
        return ''
    # Convert to string and strip any leading/trailing spaces
    zip_clean = str(zip_str).strip()
    # Take only the first 5 digits for US ZIP codes
    zip_clean = re.sub(r'[^0-9]', '', zip_clean)[:5]
    return zip_clean

# Clean ZIP codes in df1
df1_clean_zips = df1['zip_code'].apply(clean_zip_code)

# Clean ZIP codes in df2
df2_clean_zips = df2['Postal Code'].apply(clean_zip_code) if 'Postal Code' in df2.columns else pd.Series(dtype=str)

# If Postal Code column doesn't exist, alert the user
if 'Postal Code' not in df2.columns:
    print("Warning: 'Postal Code' column not found in df2")

# Match ZIP codes and store row IDs
for i, zip_code in enumerate(df1_clean_zips):
    if zip_code:  # If ZIP code is not empty
        # Find all rows where this ZIP code matches
        matching_idx = df2_clean_zips[df2_clean_zips == zip_code].index.tolist()
        
        # Store the matches
        if matching_idx:
            df1.at[i, 'ZIP_scraped_matches_row_ids'] = str(matching_idx)

# Count how many rows got matches
matched_count = df1['ZIP_scraped_matches_row_ids'].notna().sum()
print(f"ZIP code matching complete: {matched_count} rows in df1 have matches in df2")
```

```python
# STEP 3: Match by city name with permissive matching
# Create the new column to store matching row IDs
df1['City_scraped_matches_row_ids'] = None

# Function to clean and tokenize city names
def clean_city(city_str):
    if pd.isna(city_str):
        return []
    # Convert to lowercase, remove special characters
    city_clean = re.sub(r'[^a-zA-Z0-9\s]', '', str(city_str).lower()).strip()
    # Split into words
    words = city_clean.split()
    return words

# Clean and tokenize city names in df1
df1_city_tokens = df1['city'].apply(clean_city)
df1_major_city_tokens = df1['major_city'].apply(clean_city) if 'major_city' in df1.columns else pd.Series([[] for _ in range(len(df1))])

# Clean and tokenize city names in df2
if 'City' in df2.columns:
    df2_city_tokens = df2['City'].apply(clean_city)
else:
    print("Warning: 'City' column not found in df2")
    df2_city_tokens = pd.Series([[] for _ in range(len(df2))])

# Match cities using the permissive criteria
for i in range(len(df1)):
    matches = []
    
    # Combine tokens from city and major_city
    all_tokens = set(df1_city_tokens.iloc[i] + df1_major_city_tokens.iloc[i])
    
    # Skip if no tokens to match
    if not all_tokens:
        continue
    
    # For each row in df2
    for j, df2_tokens in enumerate(df2_city_tokens):
        # Check for partial matches
        if any(token in df2_tokens for token in all_tokens) or any(token in all_tokens for token in df2_tokens):
            matches.append(j)
    
    # Store the matches
    if matches:
        df1.at[i, 'City_scraped_matches_row_ids'] = str(matches)

# Count how many rows got matches
matched_count = df1['City_scraped_matches_row_ids'].notna().sum()
print(f"City matching complete: {matched_count} rows in df1 have matches in df2")
```

```python
# STEP 4: Match by exit number with permissive matching
# Create the new column to store matching row IDs
df1['Exit_scraped_matches_row_ids'] = None

# Function to extract exit numbers from text
def extract_exit_numbers(exit_str):
    if pd.isna(exit_str):
        return []
    
    # Convert to string
    exit_str = str(exit_str)
    
    # Extract numeric portions that might represent exit numbers
    exit_numbers = []
    
    # Find all numbers in the string
    number_matches = re.findall(r'\d+', exit_str)
    exit_numbers.extend(number_matches)
    
    # Special handling for compound exits like "160 EB/164 WB"
    if '/' in exit_str:
        parts = exit_str.split('/')
        for part in parts:
            part_numbers = re.findall(r'\d+', part)
            if part_numbers:
                exit_numbers.extend(part_numbers)
    
    # Remove duplicates
    return list(set(exit_numbers))

# Extract exit numbers from all relevant columns in df1
exit_columns = ['Exit_Number', 'Exit_From_Address', 'Exit_From_Label', 'Exit_Number_2', 'Exit_Number_3']
df1_exit_numbers = {}

for col in exit_columns:
    if col in df1.columns:
        df1_exit_numbers[col] = df1[col].apply(extract_exit_numbers)
    else:
        print(f"Warning: '{col}' column not found in df1")

# Extract exit numbers from df2
if 'Exit' in df2.columns:
    df2_exit_numbers = df2['Exit'].apply(extract_exit_numbers)
else:
    print("Warning: 'Exit' column not found in df2")
    df2_exit_numbers = pd.Series([[] for _ in range(len(df2))])

# Match exit numbers
for i in range(len(df1)):
    matches = []
    
    # Collect all exit numbers from this df1 row
    all_exit_numbers = []
    for col, exit_nums in df1_exit_numbers.items():
        if i < len(exit_nums):  # Ensure index is in range
            all_exit_numbers.extend(exit_nums.iloc[i])
    
    # Skip if no exit numbers to match
    if not all_exit_numbers:
        continue
    
    # For each row in df2
    for j, df2_nums in enumerate(df2_exit_numbers):
        # Check for matches
        if any(num in df2_nums for num in all_exit_numbers):
            matches.append(j)
    
    # Store the matches
    if matches:
        df1.at[i, 'Exit_scraped_matches_row_ids'] = str(matches)

# Count how many rows got matches
matched_count = df1['Exit_scraped_matches_row_ids'].notna().sum()
print(f"Exit number matching complete: {matched_count} rows in df1 have matches in df2")
```

```python
# STEP 5: Match by state
# Create the new column to store matching row IDs
df1['State_scraped_matches_row_ids'] = None

# Function to clean state names/abbreviations
def clean_state(state_str):
    if pd.isna(state_str):
        return ''
    # Convert to uppercase and remove spaces/special characters
    return re.sub(r'[^A-Z]', '', str(state_str).upper())

# Clean states in df1
df1_clean_states = df1['state'].apply(clean_state) if 'state' in df1.columns else pd.Series(dtype=str)

# Clean states in df2
df2_clean_states = df2['State'].apply(clean_state) if 'State' in df2.columns else pd.Series(dtype=str)

# If state column doesn't exist in either dataframe, alert the user
if 'state' not in df1.columns:
    print("Warning: 'state' column not found in df1")
if 'State' not in df2.columns:
    print("Warning: 'State' column not found in df2")

# Match states and store row IDs
for i, state in enumerate(df1_clean_states):
    if state:  # If state is not empty
        # Find all rows where this state matches
        matching_idx = df2_clean_states[df2_clean_states == state].index.tolist()
        
        # Store the matches
        if matching_idx:
            df1.at[i, 'State_scraped_matches_row_ids'] = str(matching_idx)

# Count how many rows got matches
matched_count = df1['State_scraped_matches_row_ids'].notna().sum()
print(f"State matching complete: {matched_count} rows in df1 have matches in df2")
```

```python
# STEP 6: Match by road name with permissive matching
# Create the new column to store matching row IDs
df1['Road_scraped_matches_row_ids'] = None

# Function to extract road identifiers and names
def extract_road_tokens(road_str):
    if pd.isna(road_str):
        return set()
    
    road_str = str(road_str).lower()
    tokens = set()
    
    # Extract basic road name
    basic_name = re.sub(r'[^a-z0-9\s]', ' ', road_str)
    tokens.update(basic_name.split())
    
    # Extract route numbers (e.g., US 60-70, I-10, NV 604)
    route_numbers = re.findall(r'(\d+)(?:\s*-\s*(\d+))?', road_str)
    for match in route_numbers:
        tokens.add(match[0])  # Add first number
        if match[1]:  # Add second number if it exists (e.g., "60-70" → add both "60" and "70")
            tokens.add(match[1])
    
    # Handle interstate notation
    interstate_match = re.search(r'i\s*[-]?\s*(\d+)', road_str)
    if interstate_match:
        tokens.add(interstate_match.group(1))  # Add just the number
    
    # Handle state routes (e.g., "NV 604" → add "604")
    state_route_match = re.search(r'[a-z]{2}\s*[-]?\s*(\d+)', road_str)
    if state_route_match:
        tokens.add(state_route_match.group(1))  # Add just the number
    
    # Handle compound routes with slashes (e.g., "NV 604/574" → add both "604" and "574")
    if '/' in road_str:
        slash_parts = road_str.split('/')
        for part in slash_parts:
            numbers = re.findall(r'\d+', part)
            tokens.update(numbers)
    
    # Extract street name without directional prefixes/suffixes (e.g., "S Main St" → "main")
    directionals = ['n', 's', 'e', 'w', 'north', 'south', 'east', 'west']
    street_types = ['st', 'ave', 'blvd', 'rd', 'ln', 'dr', 'way', 'pkwy', 'hwy', 'expwy']
    
    words = road_str.split()
    filtered_words = []
    
    for word in words:
        word = re.sub(r'[^a-z0-9]', '', word)
        if (word not in directionals and word not in street_types and 
            not (len(word) == 1 and word.isalpha())):  # Skip single letters like N, S, E, W
            filtered_words.append(word)
    
    tokens.update(filtered_words)
    
    # Remove very common words that might cause false matches
    common_words = {'the', 'of', 'and', 'to', 'a', 'in', 'for', 'is', 'on', 'that', 'by', 'this', 'with', 'i', 'you', 'it'}
    tokens = tokens - common_words
    
    return tokens

# Get road tokens from df1
road_columns = ['Main_Road', 'Secondary_Road', 'Tertiary_Road']
df1_road_tokens = {}

for col in road_columns:
    if col in df1.columns:
        df1_road_tokens[col] = df1[col].apply(extract_road_tokens)
    else:
        print(f"Warning: '{col}' column not found in df1")

# Get road tokens from df2
df2_road_columns = ['Highway', 'Street Address', 'Mailing Address', 'Road Name']
df2_road_tokens = {}

for col in df2_road_columns:
    if col in df2.columns:
        df2_road_tokens[col] = df2[col].apply(extract_road_tokens)
    else:
        print(f"Warning: '{col}' column not found in df2")

# Match roads and store row IDs
for i in range(len(df1)):
    matches = []
    
    # Collect all road tokens from this df1 row
    all_road_tokens = set()
    for col, tokens_series in df1_road_tokens.items():
        if i < len(tokens_series):  # Ensure index is in range
            all_road_tokens.update(tokens_series.iloc[i])
    
    # Skip if no road tokens to match
    if not all_road_tokens:
        continue
    
    # For each row in df2
    for j in range(len(df2)):
        # Collect all road tokens from this df2 row
        df2_all_tokens = set()
        for col, tokens_series in df2_road_tokens.items():
            if j < len(tokens_series):  # Ensure index is in range
                df2_all_tokens.update(tokens_series.iloc[j])
        
        # Check for matches
        if any(token in df2_all_tokens for token in all_road_tokens):
            matches.append(j)
    
    # Store the matches
    if matches:
        df1.at[i, 'Road_scraped_matches_row_ids'] = str(matches)

# Count how many rows got matches
matched_count = df1['Road_scraped_matches_row_ids'].notna().sum()
print(f"Road name matching complete: {matched_count} rows in df1 have matches in df2")
```

```python
# STEP 7: Match by gas station chain with permissive matching
# Create the new column to store matching row IDs
df1['Chain_scraped_matches_row_ids'] = None

# Function to standardize chain names for better matching
def standardize_chain(chain_str):
    if pd.isna(chain_str):
        return set()
    
    chain_str = str(chain_str).lower()
    tokens = set()
    
    # Standard mapping of chain variations to canonical forms
    chain_mappings = {
        'sinclair': ['sinclair', 'sinclair oil'],
        'travelcenters': ['ta', 'taexpress', 'talogo'],
        'eleven': ['11', '7-11', 'seven', 'eleven', '7 eleven', '7eleven'],
        "love's": ['love', 'loves', "love's"],
        'speedway': ['speedway', 'speedwaygas'],
        'chevron': ['chevron'],
        'flying j': ['flying j'],
        'tesoro': ['tesoro'],
        'shell': ['shell'],
        'texaco': ['texaco'],
        'exxon': ['exxon'],
        'pilot': ['pilot'],
        'conoco': ['conoco'],
        'shamrock': ['shamrock'],
        'valero': ['valero'],
        'bp': ['bp'],
        'mobil': ['mobil'],
        'circle k': ['circle_k', 'circle k'],
        'citgo': ['citgo'],
        '76': ['76', 'union 76']
    }
    
    # Add the original chain text
    tokens.add(chain_str)
    
    # Add individual words
    words = re.sub(r'[^a-z0-9\s]', ' ', chain_str).split()
    tokens.update(words)
    
    # Add known variations based on mapping
    for canonical, variations in chain_mappings.items():
        for variation in variations:
            if variation in chain_str:
                tokens.add(canonical)
                tokens.update(variations)
    
    # Specific handling for "7-Eleven" / "Eleven" because it appears in different formats
    if any(term in chain_str for term in ['11', 'seven', 'eleven']):
        tokens.update(['11', '7-11', 'seven', 'eleven', '7 eleven', '7eleven'])
    
    return tokens

# Get chain tokens from df1
if 'chain' in df1.columns:
    df1_chain_tokens = df1['chain'].apply(standardize_chain)
else:
    print("Warning: 'chain' column not found in df1")
    df1_chain_tokens = pd.Series([set() for _ in range(len(df1))])

# Get chain tokens from df2
if 'Chain' in df2.columns:
    df2_chain_tokens = df2['Chain'].apply(standardize_chain)
else:
    print("Warning: 'Chain' column not found in df2")
    df2_chain_tokens = pd.Series([set() for _ in range(len(df2))])

# Match chains and store row IDs
for i in range(len(df1)):
    df1_tokens = df1_chain_tokens.iloc[i]
    
    # Skip if no tokens to match
    if not df1_tokens:
        continue
    
    matches = []
    for j, df2_tokens in enumerate(df2_chain_tokens):
        # Check if any token matches between the two sets
        if df1_tokens.intersection(df2_tokens):
            matches.append(j)
    
    # Store the matches
    if matches:
        df1.at[i, 'Chain_scraped_matches_row_ids'] = str(matches)

# Count how many rows got matches
matched_count = df1['Chain_scraped_matches_row_ids'].notna().sum()
print(f"Chain matching complete: {matched_count} rows in df1 have matches in df2")
```

```python
# STEP 8: Match by label text
# Create the new column to store matching row IDs
df1['Label_scraped_matches_row_ids'] = None

# Function to extract meaningful words from labels
def extract_label_words(label_str):
    if pd.isna(label_str):
        return set()
    
    # Convert to lowercase and remove special characters
    clean_label = re.sub(r'[^a-z0-9\s]', ' ', str(label_str).lower())
    
    # Split into words
    words = clean_label.split()
    
    # Remove common stop words and numbers
    stop_words = {'the', '#', '&', 'and', 'or', 'a', 'an', 'at', 'by', 'for', 'from', 'in', 
                  'of', 'on', 'to', 'with', 'gas', 'station', 'service', 'store'}
    
    meaningful_words = set()
    for word in words:
        # Skip if it's just a number or a stop word
        if not word.isdigit() and word not in stop_words and len(word) > 1:
            meaningful_words.add(word)
            
            # Handle special case for "7-Eleven" type stores
            if word == '7' and any(eleven_word in words for eleven_word in ['eleven', '11']):
                meaningful_words.add('7-eleven')
            if word == 'loves':
                meaningful_words.add("love's")
    
    return meaningful_words

# Get label tokens from df1
if 'label' in df1.columns:
    df1_label_tokens = df1['label'].apply(extract_label_words)
else:
    print("Warning: 'label' column not found in df1")
    df1_label_tokens = pd.Series([set() for _ in range(len(df1))])

# Get name and chain tokens from df2
df2_name_tokens = set()
df2_chain_tokens_for_label = set()

if 'name' in df2.columns:
    df2_name_tokens = df2['name'].apply(extract_label_words)
else:
    print("Warning: 'name' column not found in df2")
    df2_name_tokens = pd.Series([set() for _ in range(len(df2))])

if 'Chain' in df2.columns:
    df2_chain_tokens_for_label = df2['Chain'].apply(extract_label_words)
else:
    print("Warning: 'Chain' column not found in df2")
    df2_chain_tokens_for_label = pd.Series([set() for _ in range(len(df2))])

# Match labels to name/chain and store row IDs
for i in range(len(df1)):
    df1_tokens = df1_label_tokens.iloc[i]
    
    # Skip if no tokens to match
    if not df1_tokens:
        continue
    
    matches = []
    for j in range(len(df2)):
        # Combine name and chain tokens from df2
        df2_combined_tokens = df2_name_tokens.iloc[j].union(df2_chain_tokens_for_label.iloc[j])
        
        # Check if any token matches
        if df1_tokens.intersection(df2_combined_tokens):
            matches.append(j)
    
    # Store the matches
    if matches:
        df1.at[i, 'Label_scraped_matches_row_ids'] = str(matches)

# Count how many rows got matches
matched_count = df1['Label_scraped_matches_row_ids'].notna().sum()
print(f"Label text matching complete: {matched_count} rows in df1 have matches in df2")
```

```python
# FINAL STEP: Summarize matches and save results
# Convert string representations of lists to actual lists
def str_to_list(str_val):
    if pd.isna(str_val):
        return []
    try:
        # Try to evaluate the string as a literal Python expression
        return eval(str_val)
    except:
        # If that fails, return an empty list
        return []

# Convert all matching columns to actual lists
match_columns = [
    'phone_scraped_matches_row_ids',
    'ZIP_scraped_matches_row_ids',
    'City_scraped_matches_row_ids',
    'Exit_scraped_matches_row_ids',
    'State_scraped_matches_row_ids',
    'Road_scraped_matches_row_ids',
    'Chain_scraped_matches_row_ids',
    'Label_scraped_matches_row_ids'
]

for col in match_columns:
    if col in df1.columns:
        df1[col] = df1[col].apply(str_to_list)

# Create a new column that combines all matches
df1['all_matches'] = df1.apply(
    lambda row: list(set().union(*[row[col] for col in match_columns if col in df1.columns and isinstance(row[col], list)])),
    axis=1
)

# Count matches by type
match_summary = {
    'Total rows in df1': len(df1),
    'Total rows in df2': len(df2)
}

for col in match_columns:
    if col in df1.columns:
        match_summary[f'Rows with {col}'] = sum(df1[col].apply(lambda x: len(x) > 0))

match_summary['Rows with any match'] = sum(df1['all_matches'].apply(lambda x: len(x) > 0))
match_summary['Rows with no matches'] = sum(df1['all_matches'].apply(lambda x: len(x) == 0))

# Calculate percentage of rows matched
match_summary['Percentage matched'] = (match_summary['Rows with any match'] / match_summary['Total rows in df1']) * 100

# Display the summary
for key, value in match_summary.items():
    if 'Percentage' in key:
        print(f"{key}: {value:.2f}%")
    else:
        print(f"{key}: {value}")

# Save the results to CSV
df1.to_csv('Add_3.csv', index=False)

# Preview sample rows with matches
print("\nSample rows with matches:")
sample_with_matches = df1[df1['all_matches'].apply(lambda x: len(x) > 0)].head(5)
display(sample_with_matches[['label', 'phone', 'city', 'state', 'chain', 'all_matches']])
```
