```python
import pandas as pd

# Load data
df = pd.read_csv(r'C:\Users\clint\Desktop\Geocoding_Task\Cleaned_Code\3_7.csv')

# Rename columns to remove OCR_ prefix
ocr_columns = [col for col in df.columns if col.startswith('OCR_')]
rename_dict = {col: col.replace('OCR_', '') for col in ocr_columns}
df = df.rename(columns=rename_dict)

#filter df for only flagged rows
#df = df[df['Flag_Reason'] == "No matches found"]
#filter for Adress_Type != 'Proper'
#df = df[df['Address_Type']  != 'Proper']
#filter for state in #filter for states CA, UT, NV, AZ
#df = df[df['state'].isin(['CA', 'UT', 'NV', 'AZ'])]
df
```

```python
# Filter for Address_Type = "Exit" and show address_standardized_OFF_parenthesis entries
exit_entries = df[df['Address_Type'] == 'Exit']
print(f"Found {len(exit_entries)} entries with Address_Type='Exit'")
print("\nEntries in address_standardized_OFF_parenthesis:")
print(exit_entries['address_standardized_OFF_parenthesis'].tolist())
```

```python
# Check which entries contain ")" characters
exit_entries_with_parenthesis = exit_entries[exit_entries['address_standardized_OFF_parenthesis'].str.contains('\)', na=False)]
print(f"Found {len(exit_entries_with_parenthesis)} entries containing ')' character:")
print("\nEntries with ')' character:")
for idx, entry in exit_entries_with_parenthesis['address_standardized_OFF_parenthesis'].items():
    print(f"Row {idx}: {entry}")
```

```python
# Remove ')' characters from address_standardized_OFF_parenthesis for Address_Type='Exit'
print("Before cleaning:")
for idx, entry in exit_entries_with_parenthesis['address_standardized_OFF_parenthesis'].items():
    print(f"Row {idx}: {entry}")

# Create a copy of the dataframe to modify
df_cleaned = df.copy()

# Remove ')' characters from entries where Address_Type='Exit'
mask = (df_cleaned['Address_Type'] == 'Exit')
df_cleaned.loc[mask, 'address_standardized_OFF_parenthesis'] = df_cleaned.loc[mask, 'address_standardized_OFF_parenthesis'].str.replace(')', '', regex=False)

print("\nAfter cleaning:")
# Show the cleaned entries
exit_entries_cleaned = df_cleaned[df_cleaned['Address_Type'] == 'Exit']
for idx in exit_entries_with_parenthesis.index:
    cleaned_entry = df_cleaned.loc[idx, 'address_standardized_OFF_parenthesis']
    print(f"Row {idx}: {cleaned_entry}")
```

```python
df= df_cleaned
df[['address_standardized_OFF_parenthesis', 'address_standardized_ON_parenthesis', 'Address_Type','zip_code','city','label']]
```
